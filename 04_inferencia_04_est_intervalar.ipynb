# Considerações Iniciais

- Em estatística, estimativas pontuais (como a média amostral) fornecem uma **única estimativa** de um parâmetro populacional.
- No entanto, toda amostra está sujeita a **variabilidade aleatória**, e confiar em apenas um valor pode ser arriscado.
- Por isso, usamos **estimativas por intervalo** para fornecer uma **faixa de valores plausíveis** para o parâmetro verdadeiro.
- Essa abordagem permite incorporar a **incerteza inerente ao processo amostral**, comunicando melhor o que sabemos (e o que não sabemos) sobre a população.


---

## Por que usar Intervalo de Confiança?

- Para refletir a **incerteza da estimativa**, que decorre do uso de uma amostra e não da população completa.
- Para **quantificar a precisão** da estimativa pontual de forma mais informativa.
- Para **auxiliar na tomada de decisões** com base em evidências estatísticas, por exemplo:
  - “Estamos **95% confiantes** de que a média de pressão arterial na população está entre 120 e 130 mmHg.”
- Para comunicar resultados de forma mais responsável e **evitar interpretações errôneas** associadas a estimativas pontuais isoladas.


---

## Definição Formal

- Um **Intervalo de Confiança (IC)** é um intervalo aleatório construído a partir dos dados amostrais que, com determinada **confiança $ (1 - \alpha) $**, contém o valor verdadeiro do parâmetro populacional.
- O nível de confiança (por exemplo, 95%) indica a **proporção de intervalos** que conterão o verdadeiro parâmetro se repetirmos o processo amostral muitas vezes.
- **Formalmente:**

$$
\mathbb{P}\left(L(X) \leq \theta \leq U(X)\right) = 1 - \alpha
$$

Onde:
- $ L(X) $ e $ U(X) $ são as extremidades inferior e superior do intervalo, funções da amostra $ X $;
- $ \theta $ é o parâmetro populacional de interesse;
- $ \alpha $ é o nível de significância (ex: 0.05 para 95% de confiança).

---

## Métodos para Construção de Intervalos

### Quantidade Pivotal

- Baseia-se em encontrar uma **função estatística conhecida**, chamada **quantidade pivotal**, que:
  - Depende dos dados e do parâmetro;
  - Tem distribuição **conhecida e independente do parâmetro**.
- Exemplo típico: $ Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}} \sim N(0,1) $
- A partir da distribuição da quantidade pivotal, constrói-se o intervalo usando quantis da distribuição.

**Vantagens:**
- Método clássico e amplamente utilizado.
- Simples quando a distribuição amostral é conhecida.

**Desvantagens:**
- Requer que certas condições sejam satisfeitas (normalidade, variância conhecida, etc.).
- Pode ser difícil de aplicar para estimadores mais complexos ou distribuições não usuais.


### Intervalo Bayesiano

- Constrói intervalos com base na **probabilidade a posteriori** de um parâmetro, assumindo que ele tem uma **distribuição prévia (prior)**.
- Ao observar os dados, atualizamos essa distribuição via o **Teorema de Bayes**, obtendo a distribuição **a posteriori**.
- Um intervalo de confiança bayesiano é chamado de **intervalo de credibilidade**.

**Formalmente:**
- Um intervalo de $ 100(1-\alpha)\%$ de credibilidade é tal que:

$$
\mathbb{P}(\theta \in [a, b] \mid \text{dados}) = 1 - \alpha
$$

**Vantagens:**
- Permite incorporar **informações prévias** ao processo inferencial.
- É intuitivo: podemos afirmar que há **95% de probabilidade de o parâmetro estar no intervalo**.

**Desvantagens:**
- Requer a escolha (muitas vezes subjetiva) de uma **distribuição a priori**.
- Pode gerar resultados diferentes dos métodos clássicos (frequentistas).

### Intervalo Bootstrap

- Método **não paramétrico**, baseado em **re-amostragem com reposição** dos dados observados.
- Em vez de usar fórmulas analíticas, simula diversas amostras para **estimar a distribuição do estimador**.

**Etapas gerais:**
1. Obtem-se a amostra original.
2. Gera-se um grande número de reamostras (com reposição).
3. Para cada reamostra, calcula-se o estimador de interesse.
4. Usa-se a distribuição empírica desses estimadores para construir o intervalo desejado.

**Vantagens:**
- Não requer suposições fortes sobre a forma da distribuição.
- Pode ser aplicado mesmo com estatísticas complexas ou sem forma analítica de variância.

**Desvantagens:**
- Computacionalmente mais custoso.
- Pode falhar com amostras muito pequenas ou com dados muito assimétricos.

## Atenção, atenção, atenção:

- O parâmetro populacional (por exemplo, a média real $\mu$) é fixo.

- O intervalo de confiança é aleatório, porque depende da amostra que você coletou.

- O que a gente quer é que, em 95% das amostras geradas, o intervalo cubra o verdadeiro valor de $\mu$.

# Intervalos de Confiança Comuns (usando quantidade pivotal)

## IC para a média com variância conhecida

### Objetivo

Estimar a média populacional $ \mu $, assumindo que a variância populacional $ \sigma^2 $ é **conhecida**.

### Quantidade Pivotal

Usamos a estatística:

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$

é uma quantidade pivotal.

### Distribuição da Quantidade Pivotal

Se os dados vêm de uma população Normal (ou amostra grande), então:

$$
Z \sim N(0,1)
$$

### Construção do Intervalo

Queremos encontrar limites **inferior** ($L$) e **superior** ($U$) tais que a probabilidade de que $ \mu $ esteja entre eles seja $1 - \alpha$:

$$
P\left( -z_{1 - \alpha/2} \leq \frac{\bar{X} - \mu}{\sigma/\sqrt{n}} \leq z_{1 - \alpha/2} \right) = 1 - \alpha
$$

Multiplicando por $ \sigma/\sqrt{n} $ e isolando $ \mu $:

$$
P\left( \bar{X} - z_{1 - \alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \leq \mu \leq \bar{X} + z_{1 - \alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \right) = 1 - \alpha
$$

### Intervalo de Confiança Final

$$
\mu \in \left[ \bar{X} \pm z_{1 - \alpha/2} \cdot \frac{\sigma}{\sqrt{n}} \right]
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parâmetros populacionais
mu = 100               # média populacional
sigma = 15             # desvio padrão populacional (conhecido)
n = 50                 # tamanho da amostra
alpha = 0.05           # nível de significância

# Gerar amostra
np.random.seed()      # aleatório a cada execução
amostra = np.random.normal(mu, sigma, n)

# Estatísticas amostrais
x_bar = np.mean(amostra)

# Quantil da normal
z = norm.ppf(1 - alpha/2)

# Erro padrão
erro = z * sigma / np.sqrt(n)

# Intervalo de confiança
ic_inf = x_bar - erro
ic_sup = x_bar + erro

print(f"Intervalo de confiança para a média (variância conhecida): [{ic_inf:.2f}, {ic_sup:.2f}]")

# Gráfico
x_vals = np.linspace(x_bar - 4*erro, x_bar + 4*erro, 1000)
y_vals = norm.pdf(x_vals, loc=x_bar, scale=sigma/np.sqrt(n))

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='Distribuição da média amostral')
plt.axvline(ic_inf, color='red', linestyle='--', label='Limite inferior do IC')
plt.axvline(ic_sup, color='red', linestyle='--', label='Limite superior do IC')
plt.axvline(mu, color='green', linestyle='-', label='Média populacional (real)')
plt.title('Intervalo de Confiança para a Média (σ conhecido)')
plt.xlabel('Valor')
plt.ylabel('Densidade')
plt.legend()
plt.grid(True)
plt.show()
```

## IC para a média com variância desconhecida

### Objetivo

Estimar a média populacional $ \mu $, assumindo que a variância populacional $ \sigma^2 $ é **desconhecida**.

### Quantidade Pivotal

Usamos a estatística:

$$
T = \frac{\bar{X} - \mu}{s / \sqrt{n}}
$$

onde $s$ é o desvio padrão amostral. Essa é uma **quantidade pivotal**.

### Distribuição da Quantidade Pivotal

Se os dados vêm de uma população Normal, então:

$$
T \sim t_{n - 1}
$$

### Construção do Intervalo

Queremos encontrar limites **inferior** ($L$) e **superior** ($U$) tais que a probabilidade de que $ \mu $ esteja entre eles seja $1 - \alpha$:

$$
P\left( -t_{1 - \alpha/2, \, n - 1} \leq \frac{\bar{X} - \mu}{s / \sqrt{n}} \leq t_{1 - \alpha/2, \, n - 1} \right) = 1 - \alpha
$$

Multiplicando por $ s/\sqrt{n} $ e isolando $ \mu $:

$$
P\left( \bar{X} - t_{1 - \alpha/2, n - 1} \cdot \frac{s}{\sqrt{n}} \leq \mu \leq \bar{X} + t_{1 - \alpha/2, n - 1} \cdot \frac{s}{\sqrt{n}} \right) = 1 - \alpha
$$

### Intervalo de Confiança Final

$$
\mu \in \left[ \bar{X} \pm t_{1 - \alpha/2, \, n - 1} \cdot \frac{s}{\sqrt{n}} \right]
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Parâmetros populacionais
mu = 100              # média populacional
sigma = 15            # desvio padrão populacional (desconhecido!)
n = 50                # tamanho da amostra
alpha = 0.05          # nível de significância

# Gerar amostra
np.random.seed()      # aleatório a cada execução
amostra = np.random.normal(mu, sigma, n)

# Estatísticas amostrais
x_bar = np.mean(amostra)
s = np.std(amostra, ddof=1)  # desvio padrão amostral

# Quantil da t de Student
t_crit = t.ppf(1 - alpha/2, df=n-1)

# Erro padrão estimado
erro = t_crit * s / np.sqrt(n)

# Intervalo de confiança
ic_inf = x_bar - erro
ic_sup = x_bar + erro

print(f"Intervalo de confiança para a média (variância desconhecida): [{ic_inf:.2f}, {ic_sup:.2f}]")

# Gráfico
x_vals = np.linspace(x_bar - 4*erro, x_bar + 4*erro, 1000)
y_vals = t.pdf((x_vals - x_bar) / (s / np.sqrt(n)), df=n-1)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='Distribuição t da média amostral')
plt.axvline(ic_inf, color='red', linestyle='--', label='Limite inferior do IC')
plt.axvline(ic_sup, color='red', linestyle='--', label='Limite superior do IC')
plt.axvline(mu, color='green', linestyle='-', label='Média populacional (real)')
plt.title('Intervalo de Confiança para a Média (σ desconhecido)')
plt.xlabel('Valor')
plt.ylabel('Densidade')
plt.legend()
plt.grid(True)
plt.show()
```

## IC para proporção

### Objetivo

Estimar a proporção populacional $p$ de uma característica de interesse (ex.: proporção de clientes satisfeitos), com base na proporção amostral $ \hat{p} $.

### Quantidade Pivotal

Usamos a estatística:

$$
Z = \frac{\hat{p} - p}{\sqrt{ \frac{ \hat{p}(1 - \hat{p}) }{n} }}
$$

Essa é uma **quantidade aproximadamente pivotal** quando $n$ é suficientemente grande.

### Distribuição da Quantidade Pivotal

Para grandes amostras (usando o Teorema Central do Limite):

$$
Z \approx N(0,1)
$$

### Construção do Intervalo

Queremos encontrar limites **inferior** ($L$) e **superior** ($U$) tais que a probabilidade de que $p$ esteja entre eles seja $1 - \alpha$:

$$
P\left( -z_{1 - \alpha/2} \leq \frac{\hat{p} - p}{\sqrt{ \frac{ \hat{p}(1 - \hat{p}) }{n} }} \leq z_{1 - \alpha/2} \right) \approx 1 - \alpha
$$

Multiplicando pelos termos do denominador e isolando $p$:

$$
P\left( \hat{p} - z_{1 - \alpha/2} \cdot \sqrt{ \frac{ \hat{p}(1 - \hat{p}) }{n} } \leq p \leq \hat{p} + z_{1 - \alpha/2} \cdot \sqrt{ \frac{ \hat{p}(1 - \hat{p}) }{n} } \right) \approx 1 - \alpha
$$

### Intervalo de Confiança Final

$$
p \in \left[ \hat{p} \pm z_{1 - \alpha/2} \cdot \sqrt{ \frac{ \hat{p}(1 - \hat{p}) }{n} } \right]
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parâmetros da amostra
n = 200                # tamanho da amostra
sucessos = 124         # número de sucessos observados
alpha = 0.05           # nível de significância

# Proporção amostral
p_hat = sucessos / n

# Quantil da normal padrão
z = norm.ppf(1 - alpha/2)

# Erro padrão
erro = z * np.sqrt(p_hat * (1 - p_hat) / n)

# Intervalo de confiança
ic_inf = p_hat - erro
ic_sup = p_hat + erro

print(f"Intervalo de confiança para a proporção: [{ic_inf:.3f}, {ic_sup:.3f}]")

# Gráfico
x_vals = np.linspace(p_hat - 4*erro, p_hat + 4*erro, 1000)
y_vals = norm.pdf(x_vals, loc=p_hat, scale=np.sqrt(p_hat * (1 - p_hat) / n))

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='Aproximação Normal')
plt.axvline(ic_inf, color='red', linestyle='--', label='Limite inferior do IC')
plt.axvline(ic_sup, color='red', linestyle='--', label='Limite superior do IC')
plt.axvline(p_hat, color='blue', linestyle='-', label='Proporção amostral')
plt.title('Intervalo de Confiança para Proporção')
plt.xlabel('Proporção')
plt.ylabel('Densidade')
plt.legend()
plt.grid(True)
plt.show()
```

## IC para a variância com média conhecida

### Objetivo

Estimar a variância populacional $ \sigma^2 $, assumindo que a média populacional $ \mu $ é **conhecida**.

### Quantidade Pivotal

Usamos a estatística:

$$
\chi^2 = \frac{ \sum_{i=1}^n (X_i - \mu)^2 }{\sigma^2}
$$

Essa é uma **quantidade pivotal**.

### Distribuição da Quantidade Pivotal

Se os dados vêm de uma população Normal com média conhecida, então:

$$
\chi^2 \sim \chi^2_n
$$

### Construção do Intervalo

Queremos encontrar limites **inferior** ($L$) e **superior** ($U$) tais que a probabilidade de que $ \sigma^2 $ esteja entre eles seja $1 - \alpha$:

$$
P\left( \chi^2_{1 - \alpha/2, \, n} \leq \frac{ \sum (X_i - \mu)^2 }{ \sigma^2 } \leq \chi^2_{\alpha/2, \, n} \right) = 1 - \alpha
$$

Isolando $ \sigma^2 $:

$$
P\left( \frac{ \sum (X_i - \mu)^2 }{ \chi^2_{\alpha/2, \, n} } \leq \sigma^2 \leq \frac{ \sum (X_i - \mu)^2 }{ \chi^2_{1 - \alpha/2, \, n} } \right) = 1 - \alpha
$$

### Intervalo de Confiança Final

$$
\sigma^2 \in \left[ \frac{ \sum (X_i - \mu)^2 }{ \chi^2_{\alpha/2, \, n} }, \, \frac{ \sum (X_i - \mu)^2 }{ \chi^2_{1 - \alpha/2, \, n} } \right]
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Parâmetros
mu = 100
sigma_real = 15
n = 30
alpha = 0.05

# Amostra simulada
np.random.seed()
x = np.random.normal(mu, sigma_real, n)

# Soma dos quadrados com média conhecida
q = np.sum((x - mu)**2)

# IC para variância
chi2_inf = chi2.ppf(1 - alpha/2, df=n)
chi2_sup = chi2.ppf(alpha/2, df=n)

ic_inf = q / chi2_inf
ic_sup = q / chi2_sup

# Valores teóricos para o gráfico (densidade da estimativa)
s2_vals = np.linspace(ic_inf * 0.8, ic_sup * 1.2, 1000)
dens_vals = chi2.pdf(q / s2_vals, df=n) * (q / (s2_vals**2))

# Valor verdadeiro da variância e estimado
s2_amostral = np.var(x, ddof=0)  # com média conhecida

print(f"Intervalo de confiança para a variância (média conhecida): [{ic_inf:.2f}, {ic_sup:.2f}]")

# Gráfico
plt.figure(figsize=(10, 5))
plt.plot(s2_vals, dens_vals, label='Densidade transformada')
plt.axvline(ic_inf, color='red', linestyle='--', label='Limite inferior do IC')
plt.axvline(ic_sup, color='red', linestyle='--', label='Limite superior do IC')
plt.axvline(sigma_real**2, color='green', linestyle='-', label='Variância populacional (real)')
plt.axvline(s2_amostral, color='blue', linestyle=':', label='Variância estimada (amostra)')
plt.title('IC para a Variância com Média Conhecida')
plt.xlabel('Valores possíveis de $\sigma^2$')
plt.ylabel('Densidade (aproximada)')
plt.grid(True)
plt.legend()
plt.show()


```

## IC para a variância com média desconhecida

### Objetivo

Estimar a variância populacional $ \sigma^2 $, assumindo que a média populacional $ \mu $ é **desconhecida**.

### Quantidade Pivotal

Usamos a estatística:

$$
\chi^2 = \frac{(n - 1) s^2}{\sigma^2}
$$

onde $s^2$ é a variância amostral. Essa é uma **quantidade pivotal**.

### Distribuição da Quantidade Pivotal

Se os dados vêm de uma população Normal, então:

$$
\chi^2 \sim \chi^2_{n - 1}
$$

### Construção do Intervalo

Queremos encontrar limites **inferior** ($L$) e **superior** ($U$) tais que a probabilidade de que $ \sigma^2 $ esteja entre eles seja $1 - \alpha$:

$$
P\left( \chi^2_{1 - \alpha/2, \, n - 1} \leq \frac{(n - 1) s^2}{\sigma^2} \leq \chi^2_{\alpha/2, \, n - 1} \right) = 1 - \alpha
$$

Isolando $ \sigma^2 $:

$$
P\left( \frac{(n - 1) s^2}{\chi^2_{\alpha/2, \, n - 1}} \leq \sigma^2 \leq \frac{(n - 1) s^2}{\chi^2_{1 - \alpha/2, \, n - 1}} \right) = 1 - \alpha
$$

### Intervalo de Confiança Final

$$
\sigma^2 \in \left[ \frac{(n - 1) s^2}{\chi^2_{\alpha/2, \, n - 1}}, \, \frac{(n - 1) s^2}{\chi^2_{1 - \alpha/2, \, n - 1}} \right]
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Parâmetros populacionais
sigma_real = 15
mu_real = 100
n = 30
alpha = 0.05

# Gerar amostra
np.random.seed()
x = np.random.normal(mu_real, sigma_real, n)

# Variância amostral com média desconhecida
s2 = np.var(x, ddof=1)  # usa (n-1)

# Intervalo de confiança para variância (média desconhecida)
chi2_inf = chi2.ppf(1 - alpha/2, df=n-1)
chi2_sup = chi2.ppf(alpha/2, df=n-1)

ic_inf = (n - 1) * s2 / chi2_inf
ic_sup = (n - 1) * s2 / chi2_sup

print(f"Intervalo de confiança para a variância (média desconhecida): [{ic_inf:.2f}, {ic_sup:.2f}]")

# Gráfico: eixo x com possíveis valores de variância
s2_vals = np.linspace(ic_inf * 0.8, ic_sup * 1.2, 1000)
q_vals = (n - 1) * s2 / s2_vals
dens_vals = chi2.pdf(q_vals, df=n - 1) * (n - 1) * s2 / (s2_vals**2)

# Gráfico
plt.figure(figsize=(10, 5))
plt.plot(s2_vals, dens_vals, label='Densidade transformada (via $\chi^2$)', color='black')
plt.axvline(ic_inf, color='red', linestyle='--', label='Limite inferior do IC')
plt.axvline(ic_sup, color='red', linestyle='--', label='Limite superior do IC')
plt.axvline(sigma_real**2, color='green', linestyle='-', label='Variância populacional (real)')
plt.axvline(s2, color='blue', linestyle=':', label='Variância amostral')
plt.title('IC para a Variância com Média Desconhecida')
plt.xlabel('Valores possíveis de $\sigma^2$')
plt.ylabel('Densidade (transformada)')
plt.grid(True)
plt.legend()
plt.show()
```

## IC para diferença de médias com variâncias conhecidas

### Objetivo

Estimar a diferença entre duas médias populacionais $ \mu_1 - \mu_2 $, assumindo que as variâncias populacionais $ \sigma_1^2 $ e $ \sigma_2^2 $ são **conhecidas**, e que as amostras são independentes.

### Quantidade Pivotal

Usamos a estatística:

$$
Z = \frac{ (\bar{X}_1 - \bar{X}_2) - (\mu_1 - \mu_2) }{ \sqrt{ \frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2} } }
$$

Essa é uma **quantidade pivotal**.

### Distribuição da Quantidade Pivotal

Se as amostras vêm de populações Normais (ou as amostras são grandes), então:

$$
Z \sim N(0,1)
$$

### Construção do Intervalo

Queremos encontrar limites **inferior** ($L$) e **superior** ($U$) tais que a probabilidade de que $ \mu_1 - \mu_2 $ esteja entre eles seja $1 - \alpha$:

$$
P\left( -z_{1 - \alpha/2} \leq \frac{ (\bar{X}_1 - \bar{X}_2) - (\mu_1 - \mu_2) }{ \sqrt{ \frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2} } } \leq z_{1 - \alpha/2} \right) = 1 - \alpha
$$

Multiplicando pelo denominador e isolando $ \mu_1 - \mu_2 $:

$$
P\left( (\bar{X}_1 - \bar{X}_2) - z_{1 - \alpha/2} \cdot \sqrt{ \frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2} } \leq \mu_1 - \mu_2 \leq (\bar{X}_1 - \bar{X}_2) + z_{1 - \alpha/2} \cdot \sqrt{ \frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2} } \right) = 1 - \alpha
$$

### Intervalo de Confiança Final

$$
\mu_1 - \mu_2 \in \left[ (\bar{X}_1 - \bar{X}_2) \pm z_{1 - \alpha/2} \cdot \sqrt{ \frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2} } \right]
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parâmetros populacionais
mu1, mu2 = 100, 105              # médias populacionais
sigma1, sigma2 = 10, 15          # desvios padrão populacionais (conhecidos)
n1, n2 = 40, 50                  # tamanhos das amostras
alpha = 0.05

# Gerar amostras independentes
np.random.seed()
x1 = np.random.normal(mu1, sigma1, n1)
x2 = np.random.normal(mu2, sigma2, n2)

# Médias amostrais
x_bar1 = np.mean(x1)
x_bar2 = np.mean(x2)
diff_xbar = x_bar1 - x_bar2

# Erro padrão da diferença
se_diff = np.sqrt(sigma1**2 / n1 + sigma2**2 / n2)

# Quantil da normal
z = norm.ppf(1 - alpha/2)

# Intervalo de confiança
erro = z * se_diff
ic_inf = diff_xbar - erro
ic_sup = diff_xbar + erro

print(f"Intervalo de confiança para a diferença de médias (variâncias conhecidas): [{ic_inf:.2f}, {ic_sup:.2f}]")

# Gráfico
x_vals = np.linspace(diff_xbar - 4*se_diff, diff_xbar + 4*se_diff, 1000)
y_vals = norm.pdf(x_vals, loc=diff_xbar, scale=se_diff)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='Distribuição da diferença amostral')
plt.axvline(ic_inf, color='red', linestyle='--', label='Limite inferior do IC')
plt.axvline(ic_sup, color='red', linestyle='--', label='Limite superior do IC')
plt.axvline(mu1 - mu2, color='green', linestyle='-', label='Diferença populacional real')
plt.axvline(diff_xbar, color='blue', linestyle=':', label='Diferença amostral')
plt.title('IC para Diferença de Médias (Variâncias Conhecidas)')
plt.xlabel('$\mu_1 - \mu_2$')
plt.ylabel('Densidade')
plt.grid(True)
plt.legend()
plt.show()
```

## IC para diferença de médias com variâncias desconhecidas mas iguais (pooled)

### Objetivo

Estimar a diferença entre duas médias populacionais $ \mu_1 - \mu_2 $, assumindo que as variâncias populacionais são **desconhecidas**, mas **iguais** ($ \sigma_1^2 = \sigma_2^2 = \sigma^2 $), com amostras independentes.

### Quantidade Pivotal

Usamos a estatística:

$$
T = \frac{ (\bar{X}_1 - \bar{X}_2) - (\mu_1 - \mu_2) }{ \sqrt{ s_p^2 \left( \frac{1}{n_1} + \frac{1}{n_2} \right) } }
$$

onde $s_p^2$ é a variância combinada (*pooled*), dada por:

$$
s_p^2 = \frac{ (n_1 - 1)s_1^2 + (n_2 - 1)s_2^2 }{ n_1 + n_2 - 2 }
$$

Essa é uma **quantidade pivotal**.

### Distribuição da Quantidade Pivotal

Sob a suposição de normalidade e variâncias iguais:

$$
T \sim t_{n_1 + n_2 - 2}
$$

### Construção do Intervalo

Queremos encontrar limites **inferior** ($L$) e **superior** ($U$) tais que a probabilidade de que $ \mu_1 - \mu_2 $ esteja entre eles seja $1 - \alpha$:

$$
P\left( -t_{1 - \alpha/2, \, n_1 + n_2 - 2} \leq \frac{ (\bar{X}_1 - \bar{X}_2) - (\mu_1 - \mu_2) }{ \sqrt{ s_p^2 \left( \frac{1}{n_1} + \frac{1}{n_2} \right) } } \leq t_{1 - \alpha/2, \, n_1 + n_2 - 2} \right) = 1 - \alpha
$$

Multiplicando pelos termos e isolando $ \mu_1 - \mu_2 $:

$$
P\left( (\bar{X}_1 - \bar{X}_2) - t_{1 - \alpha/2} \cdot \sqrt{ s_p^2 \left( \frac{1}{n_1} + \frac{1}{n_2} \right) } \leq \mu_1 - \mu_2 \leq (\bar{X}_1 - \bar{X}_2) + t_{1 - \alpha/2} \cdot \sqrt{ s_p^2 \left( \frac{1}{n_1} + \frac{1}{n_2} \right) } \right) = 1 - \alpha
$$

### Intervalo de Confiança Final

$$
\mu_1 - \mu_2 \in \left[ (\bar{X}_1 - \bar{X}_2) \pm t_{1 - \alpha/2, \, n_1 + n_2 - 2} \cdot \sqrt{ s_p^2 \left( \frac{1}{n_1} + \frac{1}{n_2} \right) } \right]
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Parâmetros populacionais
mu1, mu2 = 100, 105            # médias populacionais
sigma1, sigma2 = 12, 12        # desvios padrão populacionais (iguais, mas desconhecidos)
n1, n2 = 30, 35
alpha = 0.05

# Gerar amostras
np.random.seed()
x1 = np.random.normal(mu1, sigma1, n1)
x2 = np.random.normal(mu2, sigma2, n2)

# Médias e variâncias amostrais
x_bar1 = np.mean(x1)
x_bar2 = np.mean(x2)
s1 = np.std(x1, ddof=1)
s2 = np.std(x2, ddof=1)

# Variância combinada (pooled)
sp2 = ((n1 - 1)*s1**2 + (n2 - 1)*s2**2) / (n1 + n2 - 2)

# Erro padrão da diferença
se_pooled = np.sqrt(sp2 * (1/n1 + 1/n2))

# Quantil da t de Student
df = n1 + n2 - 2
t_crit = t.ppf(1 - alpha/2, df=df)

# Diferença amostral e intervalo
diff_xbar = x_bar1 - x_bar2
erro = t_crit * se_pooled
ic_inf = diff_xbar - erro
ic_sup = diff_xbar + erro

print(f"IC para diferença de médias (variâncias desconhecidas e iguais): [{ic_inf:.2f}, {ic_sup:.2f}]")

# Gráfico
x_vals = np.linspace(diff_xbar - 4*se_pooled, diff_xbar + 4*se_pooled, 1000)
y_vals = t.pdf((x_vals - diff_xbar)/se_pooled, df=df)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='Distribuição t da diferença amostral')
plt.axvline(ic_inf, color='red', linestyle='--', label='Limite inferior do IC')
plt.axvline(ic_sup, color='red', linestyle='--', label='Limite superior do IC')
plt.axvline(mu1 - mu2, color='green', linestyle='-', label='Diferença populacional real')
plt.axvline(diff_xbar, color='blue', linestyle=':', label='Diferença amostral')
plt.title('IC para Diferença de Médias (Variâncias Desconhecidas e Iguais)')
plt.xlabel('$\mu_1 - \mu_2$')
plt.ylabel('Densidade')
plt.grid(True)
plt.legend()
plt.show()
```

## IC para diferença de médias com variâncias desconhecidas e diferentes (Welch)

### Objetivo

Estimar a diferença entre duas médias populacionais $ \mu_1 - \mu_2 $, assumindo que as variâncias populacionais $ \sigma_1^2 $ e $ \sigma_2^2 $ são **desconhecidas e diferentes**, e que as amostras são independentes.

### Quantidade Pivotal

Usamos a estatística:

$$
T = \frac{ (\bar{X}_1 - \bar{X}_2) - (\mu_1 - \mu_2) }{ \sqrt{ \frac{s_1^2}{n_1} + \frac{s_2^2}{n_2} } }
$$

Essa é uma **quantidade pivotal**.

### Distribuição da Quantidade Pivotal

Sob a suposição de normalidade, essa estatística segue aproximadamente uma distribuição t de Student com graus de liberdade dados pela **fórmula de Welch**:

$$
T \sim t_{\nu}
$$

onde

$$
\nu = \frac{ \left( \frac{s_1^2}{n_1} + \frac{s_2^2}{n_2} \right)^2 }{ \frac{ (s_1^2 / n_1)^2 }{ n_1 - 1 } + \frac{ (s_2^2 / n_2)^2 }{ n_2 - 1 } }
$$

### Construção do Intervalo

Queremos encontrar limites **inferior** ($L$) e **superior** ($U$) tais que a probabilidade de que $ \mu_1 - \mu_2 $ esteja entre eles seja $1 - \alpha$:

$$
P\left( -t_{1 - \alpha/2, \, \nu} \leq \frac{ (\bar{X}_1 - \bar{X}_2) - (\mu_1 - \mu_2) }{ \sqrt{ \frac{s_1^2}{n_1} + \frac{s_2^2}{n_2} } } \leq t_{1 - \alpha/2, \, \nu} \right) = 1 - \alpha
$$

Multiplicando pelos termos e isolando $ \mu_1 - \mu_2 $:

$$
P\left( (\bar{X}_1 - \bar{X}_2) - t_{1 - \alpha/2, \, \nu} \cdot \sqrt{ \frac{s_1^2}{n_1} + \frac{s_2^2}{n_2} } \leq \mu_1 - \mu_2 \leq (\bar{X}_1 - \bar{X}_2) + t_{1 - \alpha/2, \, \nu} \cdot \sqrt{ \frac{s_1^2}{n_1} + \frac{s_2^2}{n_2} } \right) = 1 - \alpha
$$

### Intervalo de Confiança Final

$$
\mu_1 - \mu_2 \in \left[ (\bar{X}_1 - \bar{X}_2) \pm t_{1 - \alpha/2, \, \nu} \cdot \sqrt{ \frac{s_1^2}{n_1} + \frac{s_2^2}{n_2} } \right]
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Parâmetros populacionais
mu1, mu2 = 100, 105            # médias populacionais
sigma1, sigma2 = 10, 18        # desvios padrão populacionais (diferentes)
n1, n2 = 25, 30
alpha = 0.05

# Gerar amostras
np.random.seed()
x1 = np.random.normal(mu1, sigma1, n1)
x2 = np.random.normal(mu2, sigma2, n2)

# Estatísticas amostrais
x_bar1 = np.mean(x1)
x_bar2 = np.mean(x2)
s1 = np.std(x1, ddof=1)
s2 = np.std(x2, ddof=1)

# Diferença entre médias amostrais
diff_xbar = x_bar1 - x_bar2

# Erro padrão (não-pooled)
se_welch = np.sqrt(s1**2 / n1 + s2**2 / n2)

# Graus de liberdade de Welch
df_num = (s1**2 / n1 + s2**2 / n2)**2
df_denom = ((s1**2 / n1)**2) / (n1 - 1) + ((s2**2 / n2)**2) / (n2 - 1)
df_welch = df_num / df_denom

# Quantil t para Welch
t_crit = t.ppf(1 - alpha/2, df=df_welch)

# Intervalo de confiança
erro = t_crit * se_welch
ic_inf = diff_xbar - erro
ic_sup = diff_xbar + erro

print(f"IC para diferença de médias (Welch): [{ic_inf:.2f}, {ic_sup:.2f}]")
print(f"Graus de liberdade (Welch): {df_welch:.2f}")

# Gráfico
x_vals = np.linspace(diff_xbar - 4*se_welch, diff_xbar + 4*se_welch, 1000)
y_vals = t.pdf((x_vals - diff_xbar) / se_welch, df=df_welch)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='Distribuição t (Welch)', color='black')
plt.axvline(ic_inf, color='red', linestyle='--', label='Limite inferior do IC')
plt.axvline(ic_sup, color='red', linestyle='--', label='Limite superior do IC')
plt.axvline(mu1 - mu2, color='green', linestyle='-', label='Diferença populacional real')
plt.axvline(diff_xbar, color='blue', linestyle=':', label='Diferença amostral')
plt.title('IC para Diferença de Médias (Welch - variâncias desconhecidas e diferentes)')
plt.xlabel('$\mu_1 - \mu_2$')
plt.ylabel('Densidade')
plt.grid(True)
plt.legend()
plt.show()
```

## IC para diferença de proporções

### Objetivo

Estimar a diferença entre duas proporções populacionais $ p_1 - p_2 $, com base nas proporções amostrais $ \hat{p}_1 $ e $ \hat{p}_2 $, considerando amostras independentes.

### Quantidade Pivotal

Usamos a estatística:

$$
Z = \frac{ (\hat{p}_1 - \hat{p}_2) - (p_1 - p_2) }{ \sqrt{ \frac{ \hat{p}_1 (1 - \hat{p}_1) }{n_1} + \frac{ \hat{p}_2 (1 - \hat{p}_2) }{n_2} } }
$$

Essa é uma **quantidade aproximadamente pivotal** quando $n_1$ e $n_2$ são grandes.

### Distribuição da Quantidade Pivotal

Para amostras suficientemente grandes (TCL):

$$
Z \approx N(0,1)
$$

### Construção do Intervalo

Queremos encontrar limites **inferior** ($L$) e **superior** ($U$) tais que a probabilidade de que $ p_1 - p_2 $ esteja entre eles seja $1 - \alpha$:

$$
P\left( -z_{1 - \alpha/2} \leq \frac{ (\hat{p}_1 - \hat{p}_2) - (p_1 - p_2) }{ \sqrt{ \frac{ \hat{p}_1 (1 - \hat{p}_1) }{n_1} + \frac{ \hat{p}_2 (1 - \hat{p}_2) }{n_2} } } \leq z_{1 - \alpha/2} \right) \approx 1 - \alpha
$$

Multiplicando pelos termos do denominador e isolando $ p_1 - p_2 $:

$$
P\left( (\hat{p}_1 - \hat{p}_2) - z_{1 - \alpha/2} \cdot \sqrt{ \frac{ \hat{p}_1 (1 - \hat{p}_1) }{n_1} + \frac{ \hat{p}_2 (1 - \hat{p}_2) }{n_2} } \leq p_1 - p_2 \leq (\hat{p}_1 - \hat{p}_2) + z_{1 - \alpha/2} \cdot \sqrt{ \frac{ \hat{p}_1 (1 - \hat{p}_1) }{n_1} + \frac{ \hat{p}_2 (1 - \hat{p}_2) }{n_2} } \right)
$$

### Intervalo de Confiança Final

$$
p_1 - p_2 \in \left[ (\hat{p}_1 - \hat{p}_2) \pm z_{1 - \alpha/2} \cdot \sqrt{ \frac{ \hat{p}_1 (1 - \hat{p}_1) }{n_1} + \frac{ \hat{p}_2 (1 - \hat{p}_2) }{n_2} } \right]
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parâmetros amostrais
n1, n2 = 150, 200
sucessos1, sucessos2 = 90, 102

# Proporções amostrais
p1_hat = sucessos1 / n1
p2_hat = sucessos2 / n2
diff_p_hat = p1_hat - p2_hat

# Nível de confiança
alpha = 0.05
z = norm.ppf(1 - alpha/2)

# Erro padrão da diferença de proporções
se = np.sqrt(p1_hat * (1 - p1_hat) / n1 + p2_hat * (1 - p2_hat) / n2)

# Intervalo de confiança
erro = z * se
ic_inf = diff_p_hat - erro
ic_sup = diff_p_hat + erro

print(f"IC para diferença de proporções: [{ic_inf:.3f}, {ic_sup:.3f}]")

# Gráfico
x_vals = np.linspace(diff_p_hat - 4*se, diff_p_hat + 4*se, 1000)
y_vals = norm.pdf(x_vals, loc=diff_p_hat, scale=se)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='Distribuição aproximada (Normal)')
plt.axvline(ic_inf, color='red', linestyle='--', label='Limite inferior do IC')
plt.axvline(ic_sup, color='red', linestyle='--', label='Limite superior do IC')
plt.axvline(diff_p_hat, color='blue', linestyle=':', label='Diferença amostral')
plt.title('IC para Diferença de Proporções')
plt.xlabel('$p_1 - p_2$')
plt.ylabel('Densidade')
plt.grid(True)
plt.legend()
plt.show()
```

## IC para razão de variâncias com médias conhecidas

### Objetivo

Estimar a razão entre duas variâncias populacionais $ \frac{\sigma_1^2}{\sigma_2^2} $, assumindo que as médias populacionais $ \mu_1 $ e $ \mu_2 $ são **conhecidas**, e que as amostras são independentes.

### Quantidade Pivotal

Usamos a estatística:

$$
F = \frac{ S_1^2 }{ S_2^2 } = \frac{ \frac{1}{n_1} \sum_{i=1}^{n_1} (X_{1i} - \mu_1)^2 }{ \frac{1}{n_2} \sum_{i=1}^{n_2} (X_{2i} - \mu_2)^2 }
$$

Essa é uma **quantidade pivotal**.

### Distribuição da Quantidade Pivotal

Sob a hipótese de normalidade:

$$
F \sim F_{n_1, n_2}
$$

### Construção do Intervalo

Queremos encontrar limites **inferior** ($L$) e **superior** ($U$) tais que a probabilidade de que $ \frac{\sigma_1^2}{\sigma_2^2} $ esteja entre eles seja $1 - \alpha$:

$$
P\left( F_{1 - \alpha/2, \, n_1, \, n_2} \leq \frac{ S_1^2 }{ S_2^2 } \cdot \frac{1}{\frac{\sigma_1^2}{\sigma_2^2}} \leq F_{\alpha/2, \, n_1, \, n_2} \right) = 1 - \alpha
$$

Isolando $ \frac{\sigma_1^2}{\sigma_2^2} $:

$$
P\left( \frac{S_1^2}{S_2^2} \cdot \frac{1}{F_{\alpha/2, \, n_1, \, n_2}} \leq \frac{\sigma_1^2}{\sigma_2^2} \leq \frac{S_1^2}{S_2^2} \cdot \frac{1}{F_{1 - \alpha/2, \, n_1, \, n_2}} \right) = 1 - \alpha
$$

### Intervalo de Confiança Final

$$
\frac{\sigma_1^2}{\sigma_2^2} \in \left[ \frac{S_1^2}{S_2^2} \cdot \frac{1}{F_{\alpha/2, \, n_1, \, n_2}}, \, \frac{S_1^2}{S_2^2} \cdot \frac{1}{F_{1 - \alpha/2, \, n_1, \, n_2}} \right]
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Parâmetros populacionais
mu1, mu2 = 100, 105        # médias populacionais conhecidas
sigma1, sigma2 = 10, 15    # desvios padrão populacionais
n1, n2 = 40, 45
alpha = 0.05

# Gerar amostras
np.random.seed()
x1 = np.random.normal(mu1, sigma1, n1)
x2 = np.random.normal(mu2, sigma2, n2)

# Variâncias amostrais com médias conhecidas
s1_known = np.sum((x1 - mu1)**2) / n1  # não usa ddof
s2_known = np.sum((x2 - mu2)**2) / n2

# Estatística F
f_obs = s1_known / s2_known

# Limites da distribuição F
f_inf = f.ppf(1 - alpha/2, dfn=n1, dfd=n2)
f_sup = f.ppf(alpha/2, dfn=n1, dfd=n2)

# IC para a razão de variâncias
ic_inf = f_obs / f_inf
ic_sup = f_obs / f_sup

print(f"IC para razão de variâncias (médias conhecidas): [{ic_inf:.3f}, {ic_sup:.3f}]")

# Gráfico
r_vals = np.linspace(ic_inf * 0.6, ic_sup * 1.4, 1000)
q_vals = f_obs / r_vals
dens_vals = f.pdf(q_vals, dfn=n1, dfd=n2) * f_obs / (r_vals**2)

plt.figure(figsize=(10, 5))
plt.plot(r_vals, dens_vals, color='black', label='Densidade transformada')
plt.axvline(ic_inf, color='red', linestyle='--', label='Limite inferior do IC')
plt.axvline(ic_sup, color='red', linestyle='--', label='Limite superior do IC')
plt.axvline((sigma1**2) / (sigma2**2), color='green', linestyle='-', label='Razão real de variâncias')
plt.axvline(f_obs, color='blue', linestyle=':', label='Razão amostral')
plt.title('IC para Razão de Variâncias (Médias Conhecidas)')
plt.xlabel('$\\sigma_1^2 / \\sigma_2^2$')
plt.ylabel('Densidade')
plt.legend()
plt.grid(True)
plt.show()
```

## IC para razão de variâncias com médias desconhecidas

### Objetivo

Estimar a razão entre duas variâncias populacionais $ \frac{\sigma_1^2}{\sigma_2^2} $, assumindo que as **médias populacionais são desconhecidas**, e que as amostras são independentes.

onde $s_1^2$ e $s_2^2$ são as variâncias amostrais. Essa é uma **quantidade pivotal**.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Parâmetros populacionais
mu1, mu2 = 100, 105
sigma1, sigma2 = 10, 15
n1, n2 = 30, 35
alpha = 0.05

# Gerar amostras
np.random.seed()
x1 = np.random.normal(mu1, sigma1, n1)
x2 = np.random.normal(mu2, sigma2, n2)

# Variâncias amostrais (médias desconhecidas → usamos ddof=1)
s1 = np.var(x1, ddof=1)
s2 = np.var(x2, ddof=1)

# Estatística F observada
f_obs = s1 / s2

# Limites da distribuição F
f_sup = f.ppf(1 - alpha/2, dfn=n1 - 1, dfd=n2 - 1)
f_inf = f.ppf(alpha/2, dfn=n1 - 1, dfd=n2 - 1)

# Intervalo de confiança
ic_inf = f_obs / f_sup
ic_sup = f_obs / f_inf

print(f"IC para razão de variâncias (médias desconhecidas): [{ic_inf:.3f}, {ic_sup:.3f}]")

# Gráfico
r_vals = np.linspace(ic_inf * 0.6, ic_sup * 1.4, 1000)
q_vals = f_obs / r_vals
dens_vals = f.pdf(q_vals, dfn=n1 - 1, dfd=n2 - 1) * f_obs / (r_vals**2)

plt.figure(figsize=(10, 5))
plt.plot(r_vals, dens_vals, color='black', label='Densidade transformada')
plt.axvline(ic_inf, color='red', linestyle='--', label='Limite inferior do IC')
plt.axvline(ic_sup, color='red', linestyle='--', label='Limite superior do IC')
plt.axvline((sigma1**2)/(sigma2**2), color='green', linestyle='-', label='Razão real')
plt.axvline(f_obs, color='blue', linestyle=':', label='Razão amostral')
plt.title('IC para Razão de Variâncias (Médias Desconhecidas)')
plt.xlabel('$\\sigma_1^2 / \\sigma_2^2$')
plt.ylabel('Densidade')
plt.legend()
plt.grid(True)
plt.show()
```

# Margem de Erro

### O que é a Margem de Erro?

- A **Margem de Erro** é uma medida que expressa **o quanto uma estimativa amostral pode variar** em relação ao verdadeiro valor populacional.
- Ela representa a **metade da largura** de um intervalo de confiança simétrico.
- Em termos práticos, é o valor que adicionamos e subtraímos da estimativa pontual para construir o intervalo de confiança:

$$
\text{IC} = \text{estimativa} \pm \text{margem de erro}
$$


### Interpretação Intuitiva

- Se uma pesquisa eleitoral diz que um candidato tem **40% das intenções de voto** com margem de erro de **2%**, isso significa que:
  - O intervalo de confiança está entre **38% e 42%**.
  - Estamos **95% confiantes** (por exemplo) de que o verdadeiro apoio ao candidato está dentro desse intervalo.


### Cálculo da Margem de Erro (Exemplo Clássico)

- Em muitos casos, a margem de erro é dada por:

$$
\text{Margem de Erro} = z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
$$

Onde:
- $ z_{\alpha/2} $ é o valor crítico da distribuição normal (ex: 1,96 para 95% de confiança);
- $ \sigma $ é o desvio padrão populacional (ou uma estimativa);
- $ n $ é o tamanho da amostra.


### Fatores que Afetam a Margem de Erro

- **Tamanho da Amostra ($ n $)**:
  - Quanto maior a amostra, **menor a margem de erro**.
  - A margem de erro diminui com a **raiz quadrada** do tamanho da amostra.

- **Variabilidade dos Dados ($ \sigma $)**:
  - Quanto maior o desvio padrão, **maior a margem de erro**.
  - Dados mais dispersos tornam a estimativa menos precisa.

- **Nível de Confiança**:
  - Níveis mais altos de confiança (ex: 99%) exigem margens de erro maiores.
  - Isso ocorre porque desejamos capturar o verdadeiro parâmetro com mais certeza.


### Importância na Comunicação de Resultados

- A margem de erro ajuda a **evitar conclusões precipitadas** com base em estimativas pontuais.
- Comunica de forma transparente **a precisão da estimativa**.
- É fundamental em áreas como:
  - Pesquisas eleitorais;
  - Estudos de mercado;
  - Pesquisas científicas;
  - Controle de qualidade.


### Atenção!

- A margem de erro **não é um erro real observado**.
  - Ela é baseada em **suposições teóricas** sobre a distribuição dos dados.
- Não confunda margem de erro com erro amostral, erro padrão ou desvio padrão.


### Conclusão

- A Margem de Erro é uma ferramenta essencial para interpretar corretamente as estimativas estatísticas.
- Ela dá **contexto e credibilidade** à inferência.
- Toda vez que apresentarmos uma estimativa, devemos sempre nos perguntar:
  > **"Qual é a margem de erro?"**


```python
# 📦 Importar bibliotecas
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# 🔹 Simular população (ex: altura média de pessoas)
np.random.seed(42)
populacao = np.random.normal(loc=170, scale=10, size=100_000)

# 🔹 Sortear uma amostra aleatória
n = 100
amostra = np.random.choice(populacao, size=n, replace=False)

# 🔹 Calcular estatísticas amostrais
media_amostral = np.mean(amostra)
desvio_amostral = np.std(amostra, ddof=1)
erro_padrao = desvio_amostral / np.sqrt(n)

# 🔹 Definir nível de confiança e valor crítico Z
conf = 0.95
alpha = 1 - conf
z = stats.norm.ppf(1 - alpha / 2)

# 🔹 Calcular margem de erro e intervalo de confiança
margem_erro = z * erro_padrao
ic_inferior = media_amostral - margem_erro
ic_superior = media_amostral + margem_erro

# 🔹 Mostrar resultados
print(f"Média amostral: {media_amostral:.2f} cm")
print(f"Erro padrão: {erro_padrao:.2f} cm")
print(f"Margem de erro (95%): ±{margem_erro:.2f} cm")
print(f"Intervalo de Confiança: ({ic_inferior:.2f}, {ic_superior:.2f})")
```

```python
plt.figure(figsize=(10, 5))

# Histograma da amostra
plt.hist(amostra, bins=15, color='lightgray', edgecolor='black', alpha=0.7)

# Linha da média
plt.axvline(media_amostral, color='blue', linestyle='--', linewidth=2, label='Média Amostral')

# Linhas do intervalo de confiança
plt.axvline(ic_inferior, color='red', linestyle='--', linewidth=2, label='Limite Inferior (IC 95%)')
plt.axvline(ic_superior, color='red', linestyle='--', linewidth=2, label='Limite Superior (IC 95%)')

# Título e legenda
plt.title('Intervalo de Confiança para a Média (com Margem de Erro)')
plt.xlabel('Altura (cm)')
plt.ylabel('Frequência')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Atenção:
- A média amostral é uma estimativa do valor médio da população.
- A margem de erro representa a imprecisão associada a essa estimativa.
- O intervalo de confiança mostra uma faixa plausível para o valor real da média populacional.
- Neste caso, temos 95% de confiança de que a média verdadeira da população está entre os dois limites vermelhos.
