> **FUNDAMENTOS, APLICAO E INTERPRETAO EM DADOS DE QUESTIONRIO COM ESCALA LIKERT**

<br><br><br><br><br><br><br>

# 1. Introduo

- A **Anlise Fatorial**  uma tcnica estatstica multivariada usada para identificar **dimenses latentes** (no observadas) que explicam as correlaes entre variveis observadas.

- Essa tcnica busca **resumir um grande nmero de variveis correlacionadas** em um conjunto menor de fatores, preservando o mximo possvel da varincia comum.

-  amplamente utilizada em **psicometria**, **educao**, **marketing** e **cincias sociais** para compreender construtos subjetivos, como satisfao, engajamento, estresse e motivao, por meio de variveis observveis (itens de questionrio).

- Na prtica, a anlise fatorial permite **validar questionrios e escalas**, descobrir dimenses implcitas nos dados e calcular **escores fatoriais** para uso em outras anlises (regresso, clusterizao, etc.).


<br><br><br>


# 2. Conceito de Fatores Latentes

- Um **fator latente**  uma varivel no observvel diretamente, mas inferida a partir dos padres de correlao entre variveis observadas.

- Em psicometria, os fatores latentes representam **construtos psicolgicos**, como motivao, satisfao, engajamento ou autoeficcia.

- A Anlise Fatorial busca justamente **descobrir essas dimenses ocultas**, identificando grupos de variveis que variam de forma semelhante.

<br><br><br>


# 3. Modelo Estatstico

- Formalmente, as variveis observadas $X_1, X_2, \ldots, X_p$ podem ser representadas como combinaes lineares de $m$ fatores latentes $F_1, F_2, \ldots, F_m$:

$$
X_i = \mu_i + \lambda_{i1}F_1 + \lambda_{i2}F_2 + \cdots + \lambda_{im}F_m + \epsilon_i
$$

- onde:
  - $\mu_i$  a mdia da varivel $X_i$;
  - $\lambda_{ij}$ so as **cargas fatoriais** (correlao entre o item $i$ e o fator $j$);
  - $F_j$ so os **fatores latentes**;
  - $\epsilon_i$  o **erro especfico** da varivel $i$.
  
- A varincia total de uma varivel pode ser decomposta em duas partes:

$$
Var(X_i) = h_i^2 + \psi_i
$$

- onde:  
  - $h_i^2$ (**comunalidade**): parte da varincia de $X_i$ explicada pelos fatores latentes (quanto maior, mais o item participa do fator)
  - $\psi_i$ (**unicidade**): parte da varincia exclusiva de $X_i$, no compartilhada com outros itens (erro ou especificidade).

- O propsito da anlise fatorial  decompor a matriz de correlao dos dados da seguinte forma:
$$
R \approx \Lambda \Lambda' + \Psi
$$

- onde:
  - $R$: matriz de correlao entre as variveis observadas
  - $\Lambda$: matriz de cargas fatoriais
  - $\Psi$: matriz diagonal das unicidades

- Ou seja, a ideia  **estimar $\Lambda$ e $\Psi$** de modo que essa decomposio reproduza a estrutura de correlaes observada entre as variveis.

<br><br><br>


# 4. Etapas Principais da Anlise Fatorial

- Avaliao de Adequao dos Dados
  - Teste de **Bartlett**  verifica se a matriz de correlaes difere da identidade.  
  - **KMO (Kaiser-Meyer-Olkin)**  mede a proporo de varincia comum entre variveis.  
    - KMO > 0.80  timo  
    - 0.700.79  bom  
    - 0.600.69  aceitvel  
    - < 0.50  inadequado

- Determinao do Nmero de Fatores
  - Critrios mais usados:  
    - **Autovalor > 1 (Kaiser)**  
    - **Scree Plot** (identificao do cotovelo)  
    - **Parallel Analysis** (autovalores reais > aleatrios)

- Extrao dos Fatores
  - Mtodos principais:  
    - *Principal Axis Factoring (PAF)*  
    - *Maximum Likelihood (ML)*  

- Rotao dos Fatores
  - O objetivo  tornar as cargas mais interpretveis.  
  - **Ortogonal (Varimax)**  fatores independentes.  
  - **Oblqua (Promax, Oblimin)**  fatores podem se correlacionar.

- Interpretao dos Fatores
  - Cargas  0.40 so consideradas relevantes.  
  - Variveis com cargas elevadas em um mesmo fator formam uma **dimenso comum**.  
  - A **comunalidade ($h_i^2$)** indica quanto da varincia do item foi explicada pelo modelo.


<br><br><br>


# 5. Diferena entre Anlise Fatorial e PCA

| Aspecto | Anlise Fatorial | PCA |
|----------|------------------|-----|
| Objetivo | Identificar fatores latentes | Reduzir varincia total |
| Varincia usada | Apenas a comum | Toda a varincia |
| Modelo terico | Estrutural (modelo causal) | Algbrico (ortogonalizao) |
| Interpretao | Construtos latentes (ex: engajamento) | Componentes sem interpretao psicolgica |

# 6. Aplicao Prtica: Questionrio Likert (Satisfao e Engajamento)

<br>

## Descrio da Base de Dados

- Base com **8 itens em escala Likert (15)**, respondidos por **600 indivduos**.
- Lista de perguntas:  
  - Q1: Estou satisfeito com meu trabalho atual.  
  - Q2: Sinto orgulho de fazer parte desta empresa.  
  - Q3: As tarefas que realizo so interessantes e me motivam.  
  - Q4: Sinto que o meu trabalho  reconhecido pela organizao.  
  - Q5: Sinto-me entusiasmado ao iniciar o meu dia de trabalho.  
  - Q6: Tenho energia e disposio para alcanar as metas da empresa.  
  - Q7: Estou totalmente envolvido com as atividades do meu time.  
  - Q8: Quando trabalho, sinto que estou contribuindo para algo importante.
- Os itens foram elaborados para medir dois construtos psicolgicos:  
  - **Fator 1  Satisfao no Trabalho** (Q1Q4)  
  - **Fator 2  Engajamento Organizacional** (Q5Q8)



<br>

## Procedimentos

- Gerar a base sinttica (distribuio realista de respostas 15).  
- Calcular **KMO** e **Teste de Bartlett** (adequao da matriz de correlaes).  
- Aplicar **Parallel Analysis** para definir o nmero de fatores.  
- Rodar **Anlise Fatorial (ML)** com rotao **Varimax**.  
- Examinar **cargas fatoriais**, **comunalidades** e **varincia explicada**.  
- Calcular **escores fatoriais** para cada indivduo (Satisfao e Engajamento).

<br>

## Usando Python

```python
# Instalao e imports
!pip install factor_analyzer pandas numpy matplotlib scikit-learn --quiet

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from factor_analyzer.factor_analyzer import (
    calculate_kmo,
    calculate_bartlett_sphericity,
    FactorAnalyzer
)

np.random.seed(7)
```

```python
# Gerao do questionrio Likert sinttico
# - 8 itens, Likert 15
# - Dois construtos: Satisfao (Q1Q4) e Engajamento (Q5Q8)

n = 600  # nmero de respondentes

# Fatores latentes "verdadeiros"
F1 = np.random.normal(size=n)  # Satisfao
F2 = np.random.normal(size=n)  # Engajamento

# Matriz de cargas "verdadeiras" (8 itens x 2 fatores)
# Q1Q4 saturam em Satisfao; Q5Q8 saturam em Engajamento
L_true = np.array([
    [0.82, 0.10],  # Q1
    [0.78, 0.08],  # Q2
    [0.75, 0.12],  # Q3
    [0.70, 0.05],  # Q4
    [0.05, 0.80],  # Q5
    [0.08, 0.77],  # Q6
    [0.10, 0.74],  # Q7
    [0.12, 0.70]   # Q8
])

# Erros especficos
E = np.random.normal(scale=0.35, size=(n, 8))

# Respostas contnuas (modelo linear latente)
Y_cont = np.vstack([F1, F2]).T @ L_true.T + E  # (n x 8)

def to_likert(col_cont):
    """Discretiza uma coluna contnua para Likert 15 via quintis."""
    q = np.quantile(col_cont, [0.2, 0.4, 0.6, 0.8])
    return np.digitize(col_cont, q) + 1

# DataFrame Likert
df = pd.DataFrame({f"Q{i+1}": to_likert(Y_cont[:, i]) for i in range(8)})
df.head(10)
```

```python
# Adequao  Anlise Fatorial: KMO e Bartlett

# KMO
# - avalia se as variveis tm correlao suficiente entre si para formar fatores.
# - se as correlaes forem muito baixas (ou seja, cada varivel se comporta isoladamente), a anlise fatorial no faz sentido.
kmo_all, kmo_model = calculate_kmo(df)


# Bartlett
# - avalia se a matriz de correlaes  significativamente diferente da matriz identidade
# - verifica se h correlao suficiente entre as variveis para justificar uma anlise fatorial
# - a hiptese nula (H_0)  que a matriz de correlaes  a identidade (ou seja, no h correlaes relevantes)
chi2, pval = calculate_bartlett_sphericity(df)

# Sadas:
print(f"KMO (global): {kmo_model:.3f}")
print(f"Bartlett: chi2={chi2:.2f}, p-value={pval:.3e}")

# Regras prticas:
# - KMO >= 0.70  bom; >= 0.80  timo
# - p-valor do Bartlett < 0.05 sugere correlao suficiente entre itens.
```

```python
# Definio do nmero de fatores

# Podemos usar:
# - Kaiser (autovalor > 1) costuma superestimar o nmero de fatores
# - Scree plot  subjetivo (depende do olho)
# - Parallel Analysis:  emprica e comparativa (responde meu autovalor  maior do que esperaramos por puro rudo?

# Em psicometria, PA  recomendao moderna para deciso de reteno.
# Compara autovalores dos dados vs. autovalores de dados aleatrios
# Como funciona:
# - A PA gera vrias amostras aleatrias com mesma dimenso (n e p)
# - E, para cada uma, calcula os autovalores
# - Depois, ela tira a mdia desses autovalores para criar a linha de referncia
# Se os dois primeiros autovalores reais esto acima da linha aleatria e o terceiro j cai abaixo, voc retenha 2 fatores.


def parallel_analysis(X, n_iter=500, random_state=42):
    rng = np.random.default_rng(random_state)
    X = np.asarray(X)
    n, p = X.shape

    # Padroniza
    Z = (X - X.mean(axis=0)) / X.std(axis=0, ddof=1)

    # Autovalores dos dados
    eigvals_data = np.linalg.eigvalsh(np.cov(Z, rowvar=False))
    eigvals_data = np.sort(eigvals_data)[::-1]

    # Autovalores mdios de dados aleatrios com mesma dimenso
    rand_eigs = np.zeros((n_iter, p))
    for i in range(n_iter):
        R = rng.normal(size=(n, p))
        R = (R - R.mean(axis=0)) / R.std(axis=0, ddof=1)
        rand_eigs[i, :] = np.sort(np.linalg.eigvalsh(np.cov(R, rowvar=False)))[::-1]
    mean_rand = rand_eigs.mean(axis=0)
    return eigvals_data, mean_rand

eigvals_data, eigvals_rand = parallel_analysis(df.values, n_iter=500)

plt.figure()
plt.plot(range(1, len(eigvals_data)+1), eigvals_data, marker='o', label='Autovalores (dados)')
plt.plot(range(1, len(eigvals_rand)+1), eigvals_rand, marker='o', label='Autovalores mdios (aleatrios)')
plt.xlabel('Componente')
plt.ylabel('Autovalor')
plt.title('Parallel Analysis  Questionrio Likert')
plt.legend()
plt.tight_layout()
plt.show()

n_factors = int(np.sum(eigvals_data > eigvals_rand))
n_factors = max(1, n_factors)
print("Nmero sugerido de fatores:", n_factors)
```

```python
# Anlise Fatorial Exploratria
# - momento de extrair e interpretar os fatores latentes
```

```python
# Ajuste do modelo

fa = FactorAnalyzer(n_factors=n_factors, method='ml', rotation='varimax')
fa.fit(df)
# n_factors=n_factors: o nmero de fatores a extrair (definido na Parallel Analysis).
# method='ml': usa o mtodo da Mxima Verossimilhana
# rotation='varimax': aplica a rotao Varimax, que  ortogonal (mantm os fatores no correlacionados).
```

```python
# Cargas fatoriais rotacionadas (lambda_{ij})
# - So as correlaes entre as variveis observadas e os fatores latentes
# - Indicam quanto uma varivel participa daquele fator
# - Cada varivel (linha) tem uma carga para cada fator (coluna)

loadings = pd.DataFrame(
    fa.loadings_,
    index=df.columns,
    columns=[f"F{i+1}" for i in range(n_factors)]
).round(3)
loadings

# Regra prtica:
# -  Cargas > 0.40 so consideradas relevantes.
# - Cargas > 0.70 so muito fortes.
# - Se uma varivel carrega alto em dois fatores, ela pode estar ambgua  possvel excluso ou reviso do item.
```

```python
# Grfico de barras para as cargas fatoriais
import seaborn as sns
loadings_abs = loadings.abs().reset_index().melt(id_vars='index', var_name='Fator', value_name='|Carga|')
plt.figure(figsize=(8,5))
sns.barplot(data=loadings_abs, x='index', y='|Carga|', hue='Fator')
plt.title('Cargas fatoriais por varivel')
plt.xlabel('Itens do questionrio')
plt.ylabel('|Carga fatorial|')
plt.legend(title='Fator')
plt.tight_layout()
plt.show()
```

```python
# Comunalidades
# - Cada comunalidade ($h_i^2$) representa a proporo da varincia total da varivel que  explicada pelos fatores.
# -  calculada como a soma dos quadrados das cargas daquela varivel nos fatores retidos.
communalities = pd.Series(fa.get_communalities(), index=df.columns, name="Comunalidade").round(3)
communalities.to_frame()

# Interpretao:
# - Q1 tem 81% de sua varincia explicada pelos fatores: tima representatividade.

# Regra prtica:
# - Comunalidades < 0.4 indicam que o item no se encaixa bem na estrutura fatorial: pode ser revisado.
```

```python
# Varincia por fator
# - Varincia absoluta explicada por cada fator
# - Proporo da varincia (frao total explicada)
# - Proporo acumulada
var_explained, prop_var, cum_var = fa.get_factor_variance()
variance_info = pd.DataFrame({
    "Varincia": np.round(var_explained, 3),
    "Proporo": np.round(prop_var, 3),
    "Acumulada": np.round(cum_var, 3)
}, index=[f"F{i+1}" for i in range(n_factors)])
variance_info
```

```python
# Escores fatoriais por respondente
# - Mostrar quanto cada pessoa tem desses fatores
scores = pd.DataFrame(
    fa.transform(df),
    columns=[f"score_F{i+1}" for i in range(n_factors)]
)
out = pd.concat([df, scores], axis=1)
out.head(10)

# Intepretao:
# - indivduo 0:
#   - score_F1 = +0.16  ligeiramente acima da mdia em Satisfao.
#   - score_F2 = +1.49  muito acima da mdia em Engajamento.
#   - Perfil: entusiasmado e envolvido, mas satisfao apenas moderada.
# - indivduo 2:
#   - score_F1 = -1.09
#   - score_F2 = -0.48
#   - Perfil: pouco satisfeito, pouco engajado.

```

```python
# Confiabilidade interna ( de Cronbach)
# -  um ndice de consistncia interna
# - Mede quanto os itens de uma escala esto correlacionados entre si
# - Se todos os itens realmente medem o mesmo conceito latente, eles tendem a variar juntos  e o alfa ser alto

def cronbach_alpha(df_items):
    # df_items: somente as colunas dos itens de uma escala
    k = df_items.shape[1]
    item_vars = df_items.var(axis=0, ddof=1).sum()
    total_var = df_items.sum(axis=1).var(ddof=1)
    return (k/(k-1)) * (1 - item_vars/total_var)

itens_satisfacao   = ["Q1","Q2","Q3","Q4"]
itens_engajamento  = ["Q5","Q6","Q7","Q8"]

alpha_sat = cronbach_alpha(df[itens_satisfacao])
alpha_eng = cronbach_alpha(df[itens_engajamento])

print(f"Alfa de Cronbach  Satisfao (Q1Q4): {alpha_sat:.3f}")
print(f"Alfa de Cronbach  Engajamento (Q5Q8): {alpha_eng:.3f}")

# Regra prtica (no rgida):
#   > 0.70 = aceitvel, > 0.80 = bom, > 0.90 = excelente (pode indicar redundncia)
```
