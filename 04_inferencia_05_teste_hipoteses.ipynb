# Por que usar um Teste de Hipóteses?
- **Motivação:** tomada de decisão sob incerteza.
- Permite avaliar se uma evidência observada é consistente com uma hipótese inicial (H₀) ou sugere uma alternativa (H₁).
- **Exemplos clássicos:**  
  - **Problema da decisão do júri**:
    - H₀: o réu é inocente
    - H₁: o réu é culpado
    - O júri deve decidir baseado em evidências, assumindo o risco de cometer erros
  -  **Detecção de fraude em transações:**
    - H₀: a transação é legítima
    - H₁: a transação é fraudulenta
    - O sistema antifraude decide aprovar ou bloquear a transação, assumindo o risco de bloquear uma compra legítima (erro tipo I) ou aprovar uma compra fraudulenta (erro tipo II)
  - **Detecção de vírus em teste laboratorial:**
    - H₀: a amostra não contém o vírus
    - H₁: a amostra contém o vírus
    - O teste diagnóstico decide reportar positivo ou negativo, assumindo o risco de falso positivo (erro tipo I) ou falso negativo (erro tipo II)


---

# Definição Formal

- Um **Teste de Hipóteses** é um procedimento estatístico utilizado para decidir, com base em dados amostrais, se aceitamos ou rejeitamos uma **hipótese nula** sobre um parâmetro populacional.  
- A decisão é tomada a partir de uma **estatística de teste** e de uma **região de rejeição**, controlando a probabilidade de cometer erros.  
- **Formalmente:**  

$$
\phi(x) =
\begin{cases}
1, & \text{rejeitar } H_0 \quad (\text{aceitar } H_1) \\
0, & \text{não rejeitar } H_0
\end{cases}
$$

Onde:  
- $H_0$: hipótese nula (afirmação inicial sobre o parâmetro populacional);  
- $H_1$: hipótese alternativa (afirmação oposta ou complementar a $H_0$);  
- $\phi(x)$: **função de decisão**, que retorna 1 se rejeitamos $H_0$ e 0 caso contrário;  
- **Erro Tipo I ($\alpha$)**: $\mathbb{P}(\text{rejeitar } H_0 \mid H_0 \ \text{verdadeira})$;  
- **Erro Tipo II ($\beta$)**: $\mathbb{P}(\text{não rejeitar } H_0 \mid H_0 \ \text{falsa})$;  
- **Região de aceitação**: conjunto de valores da estatística de teste que levam a **não rejeitar** $H_0$;  
- **Região de rejeição**: conjunto de valores da estatística de teste que levam a **rejeitar** $H_0$.

# Testes de Hipóteses Comuns (usando quantidade pivotal)

## Teste de Hipóteses para a média com variância conhecida

### Objetivo

Testar se a média populacional $ \mu $ é igual a um valor específico $ \mu_0 $, assumindo que a variância populacional $ \sigma^2 $ é **conhecida**.

### Hipóteses

- **Bicaudal**:
$$
H_0: \mu = \mu_0 \quad \text{vs} \quad H_1: \mu \neq \mu_0
$$

- **Unicaudal à direita**:
$$
H_0: \mu \leq \mu_0 \quad \text{vs} \quad H_1: \mu > \mu_0
$$

- **Unicaudal à esquerda**:
$$
H_0: \mu \geq \mu_0 \quad \text{vs} \quad H_1: \mu < \mu_0
$$

### Quantidade Pivotal

Usamos a estatística:

$$
Z = \frac{\bar{X} - \mu_0}{\sigma / \sqrt{n}}
$$

que é uma quantidade pivotal.

### Distribuição da Quantidade Pivotal

Se os dados vêm de uma população Normal (ou amostra grande), então:

$$
Z \sim N(0,1) \quad \text{sob } H_0
$$

### Construção do Teste

- **Bicaudal**: rejeitar $H_0$ se
$$
|Z_{\text{obs}}| > z_{1 - \alpha/2}
$$

- **Unicaudal à direita**: rejeitar $H_0$ se
$$
Z_{\text{obs}} > z_{1 - \alpha}
$$

- **Unicaudal à esquerda**: rejeitar $H_0$ se
$$
Z_{\text{obs}} < -z_{1 - \alpha}
$$

### Regra de Decisão (bicaudal)

1. Calcular $Z_{\text{obs}}$ com a amostra.
2. Obter o valor crítico $z_{1 - \alpha/2}$.
3. **Rejeitar $H_0$** se $|Z_{\text{obs}}| > z_{1 - \alpha/2}$.

### Teste de Hipóteses Final

- Estatística:
$$
Z_{\text{obs}} = \frac{\bar{X} - \mu_0}{\sigma / \sqrt{n}}
$$

- Decisão:
  - Se $|Z_{\text{obs}}| > z_{1 - \alpha/2}$ → **Rejeita $H_0$**.
  - Caso contrário → **Não rejeita $H_0$**.


```python
# -----------------------------------------------------------
# Figura: regiões de aceitação e rejeição em testes Z ou t
# -----------------------------------------------------------
# Parâmetros para editar:
TEST_TYPE = "two-sided"   # "two-sided", "right", "left"
DISTRIBUTION = "z"        # "z" ou "t"
alpha = 0.05              # nível de significância
df = 14                   # g.l. se DISTRIBUTION == "t"
x_obs = None              # por ex.: 2.1 para marcar a estatística observada

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats as st

# Escolhe distribuição e suporte
if DISTRIBUTION == "z":
    dist = st.norm
    x_min, x_max = -4.2, 4.2
    pdf = dist.pdf
    ppf = dist.ppf
    label = "N(0,1)"
elif DISTRIBUTION == "t":
    dist = st.t(df=df)
    x_min, x_max = -5.5, 5.5
    pdf = dist.pdf
    ppf = dist.ppf
    label = f"t(df={df})"
else:
    raise ValueError("DISTRIBUTION deve ser 'z' ou 't'.")

x = np.linspace(x_min, x_max, 2000)
y = pdf(x) if DISTRIBUTION == "z" else dist.pdf(x)

# Quantis críticos e regiões
if TEST_TYPE == "two-sided":
    c_left  = ppf(alpha/2)
    c_right = ppf(1 - alpha/2)
    accept_region = (x >= c_left) & (x <= c_right)
    rej_left  = x < c_left
    rej_right = x > c_right
    title = f"{label} — Teste bilateral (α={alpha})"
elif TEST_TYPE == "right":
    c_right = ppf(1 - alpha)
    accept_region = x <= c_right
    rej_left  = np.zeros_like(x, dtype=bool)
    rej_right = x > c_right
    title = f"{label} — Teste unicaudal à direita (α={alpha})"
elif TEST_TYPE == "left":
    c_left = ppf(alpha)
    accept_region = x >= c_left
    rej_left  = x < c_left
    rej_right = np.zeros_like(x, dtype=bool)
    title = f"{label} — Teste unicaudal à esquerda (α={alpha})"
else:
    raise ValueError("TEST_TYPE deve ser 'two-sided', 'right' ou 'left'.")

# Plot
plt.figure(figsize=(10, 5))
plt.plot(x, y, linewidth=2, label=f"Densidade {label}")

# Sombreamentos
plt.fill_between(x, 0, y, where=accept_region, alpha=0.25, label="Região de ACEITAÇÃO (não rejeita H0)")
if rej_left.any():
    plt.fill_between(x, 0, y, where=rej_left, alpha=0.35, label="Região de REJEIÇÃO (cauda esquerda)")
if rej_right.any():
    plt.fill_between(x, 0, y, where=rej_right, alpha=0.35, label="Região de REJEIÇÃO (cauda direita)")

# Linhas críticas e rótulos
if TEST_TYPE in ("two-sided", "left"):
    y_cl = pdf(c_left) if DISTRIBUTION == "z" else dist.pdf(c_left)
    plt.axvline(c_left, linestyle="--", linewidth=1.5)
    plt.text(c_left, y_cl, f"  cₗ = {c_left:.3f}", rotation=90, va="bottom", ha="left")

if TEST_TYPE in ("two-sided", "right"):
    y_cr = pdf(c_right) if DISTRIBUTION == "z" else dist.pdf(c_right)
    plt.axvline(c_right, linestyle="--", linewidth=1.5)
    plt.text(c_right, y_cr, f"  cᵣ = {c_right:.3f}", rotation=90, va="bottom", ha="left")

# Estatística observada (opcional)
if x_obs is not None:
    y_obs = pdf(x_obs) if DISTRIBUTION == "z" else dist.pdf(x_obs)
    plt.axvline(x_obs, linewidth=2)
    plt.text(x_obs, y_obs, f"  z/t_obs = {x_obs:.3f}", rotation=90, va="bottom", ha="left")

# Anotação do alfa
if TEST_TYPE == "two-sided":
    plt.text(ppf(0.03), max(y)*0.85, f"Áreas de rejeição: α/2 = {alpha/2:.3f} em cada cauda")
elif TEST_TYPE == "right":
    plt.text(ppf(0.02), max(y)*0.85, f"Área de rejeição à direita: α = {alpha:.3f}")
else:
    plt.text(ppf(0.55), max(y)*0.85, f"Área de rejeição à esquerda: α = {alpha:.3f}")

plt.title(title)
plt.xlabel("Estatística de teste")
plt.ylabel("Densidade")
plt.legend(loc="upper center", ncol=2)
plt.tight_layout()
plt.show()
```

## Teste de Hipóteses para a média com variância desconhecida

### Objetivo

Testar se a média populacional $ \mu $ é igual a um valor específico $ \mu_0 $, assumindo que a variância populacional $ \sigma^2 $ é **desconhecida**.

### Hipóteses

- **Bicaudal**:
$$
H_0: \mu = \mu_0 \quad \text{vs} \quad H_1: \mu \neq \mu_0
$$

- **Unicaudal à direita**:
$$
H_0: \mu \leq \mu_0 \quad \text{vs} \quad H_1: \mu > \mu_0
$$

- **Unicaudal à esquerda**:
$$
H_0: \mu \geq \mu_0 \quad \text{vs} \quad H_1: \mu < \mu_0
$$

### Quantidade Pivotal

Usamos a estatística:

$$
T = \frac{\bar{X} - \mu_0}{S / \sqrt{n}}
$$

onde:
- $\bar{X}$ é a média amostral;
- $S$ é o desvio-padrão amostral.

### Distribuição da Quantidade Pivotal

Se os dados vêm de uma população Normal, então:

$$
T \sim t_{n-1} \quad \text{sob } H_0
$$

onde $t_{n-1}$ é a distribuição t de Student com $n - 1$ graus de liberdade.

### Construção do Teste

- **Bicaudal**: rejeitar $H_0$ se
$$
|T_{\text{obs}}| > t_{1 - \alpha/2, \, n-1}
$$

- **Unicaudal à direita**: rejeitar $H_0$ se
$$
T_{\text{obs}} > t_{1 - \alpha, \, n-1}
$$

- **Unicaudal à esquerda**: rejeitar $H_0$ se
$$
T_{\text{obs}} < -t_{1 - \alpha, \, n-1}
$$

### Regra de Decisão (bicaudal)

1. Calcular $T_{\text{obs}}$ com a amostra.
2. Obter o valor crítico $t_{1 - \alpha/2, \, n-1}$.
3. **Rejeitar $H_0$** se $|T_{\text{obs}}| > t_{1 - \alpha/2, \, n-1}$.

### Teste de Hipóteses Final

- Estatística:
$$
T_{\text{obs}} = \frac{\bar{X} - \mu_0}{S / \sqrt{n}}
$$

- Decisão:
  - Se $|T_{\text{obs}}| > t_{1 - \alpha/2, \, n-1}$ → **Rejeita $H_0$**.
  - Caso contrário → **Não rejeita $H_0$**.

# Métodos para Avaliação de Testes de Hipóteses
- **Nível de significância (α)** — controle do erro tipo I.
- **Poder do teste (1 - β)** — probabilidade de rejeitar H₀ quando ela é falsa.
- **Tamanho do efeito** — magnitude prática da diferença detectada.
- **Curvas ROC e área sob a curva (AUC)** — especialmente em classificação binária.
- **Robustez** — sensibilidade do teste a desvios das suposições.

# Os Testes Hipóteses mais usados

```python
# ---------------------------------------------------------------------------
#  Testes de Normalidade (paramétricos ou baseados em momentos)
# ---------------------------------------------------------------------------
# Servem para verificar se os dados seguem uma distribuição Normal — pré-requisito
# comum para testes paramétricos de média, variância e ANOVA.

import numpy as np
import pandas as pd
from scipy import stats as st

# ---------------------------------------------------------------------------
# Gerando a base:
# ---------------------------------------------------------------------------
rng = np.random.default_rng(42)
n = 200
alpha = 0.05
x_norm  = rng.normal(loc=0.0, scale=1.0, size=n)# Dados que seguem a Normal(0,1) — esperado "passar" nos testes
x_heavy = rng.standard_t(df=3, size=n)# Dados com caudas pesadas (t de Student com 3 g.l.) — esperado "reprovar"
X = x_norm      # mude para x_heavy para ver o contraste
mu0, sigma0 = 0.0, 1.0# Parâmetros conhecidos para KS paramétrico

# ---------------------------------------------------------------------------
# Shapiro–Wilk
# ---------------------------------------------------------------------------
## Usado quando: Amostras pequenas a médias (até ~5000 no SciPy).
## Vantagem: Muito eficiente para detectar não-normalidade em n pequeno/médio.
## Desvantagem: Em n muito grande, qualquer desvio mínimo vira significante.
stat_shapiro, p_shapiro = st.shapiro(X)
print("Shapiro–Wilk:")
print(f"  Estatística = {stat_shapiro:.4f}, p-valor = {p_shapiro:.4f}")
if p_shapiro < alpha:
    print("  Decisão: Rejeita H0 (não Normal)\n")
else:
    print("  Decisão: Não rejeita H0 (Normal)\n")

# ---------------------------------------------------------------------------
# Anderson–Darling
# ---------------------------------------------------------------------------
## Usado quando: Sensível a desvios nas caudas da distribuição.
## Vantagem: Fornece valores críticos para vários níveis de significância.
## Desvantagem: Não retorna p-valor no SciPy.
res_ad = st.anderson(X, dist='norm')
print("Anderson–Darling:")
print(f"  Estatística = {res_ad.statistic:.4f}")
for sl, cv in zip(res_ad.significance_level, res_ad.critical_values):
    print(f"  Nível {sl:.1f}% → valor crítico = {cv:.4f}")
# Comparação com α definido
idx_alpha = (np.abs(np.array(res_ad.significance_level)/100 - alpha)).argmin()
cv_alpha = res_ad.critical_values[idx_alpha]
if res_ad.statistic > cv_alpha:
    print(f"  Decisão (α={alpha}): Rejeita H0 (não Normal)\n")
else:
    print(f"  Decisão (α={alpha}): Não rejeita H0 (Normal)\n")

# ---------------------------------------------------------------------------
# Kolmogorov–Smirnov (com parâmetros conhecidos)
# ---------------------------------------------------------------------------
## Usado quando: μ e σ populacionais são conhecidos a priori.
## Vantagem: Teste exato contra N(μ0, σ0).
## Desvantagem: Não válido se estimar μ e σ a partir dos dados (nesse caso, usar Lilliefors).
stat_ks, p_ks = st.kstest(X, 'norm', args=(mu0, sigma0))
print("Kolmogorov–Smirnov (μ,σ conhecidos):")
print(f"  Estatística = {stat_ks:.4f}, p-valor = {p_ks:.4f}")
if p_ks < alpha:
    print("  Decisão: Rejeita H0 (não Normal)\n")
else:
    print("  Decisão: Não rejeita H0 (Normal)\n")

# ---------------------------------------------------------------------------
# Jarque–Bera
# ---------------------------------------------------------------------------
## Usado quando: Quer verificar normalidade via assimetria e curtose.
## Vantagem: Simples, rápido; muito usado em econometria.
## Desvantagem: Assintótico — menos confiável em amostras pequenas.
stat_jb, p_jb = st.jarque_bera(X)
print("Jarque–Bera:")
print(f"  Estatística = {stat_jb:.4f}, p-valor = {p_jb:.4f}")
if p_jb < alpha:
    print("  Decisão: Rejeita H0 (não Normal)\n")
else:
    print("  Decisão: Não rejeita H0 (Normal)\n")

# ---------------------------------------------------------------------------
# D’Agostino–Pearson (Omnibus)
# ---------------------------------------------------------------------------
## Usado quando: Quer um teste "omnibus" combinando skewness e kurtosis.
## Vantagem: Bom poder para vários tipos de desvio; funciona bem com n ≥ 20.
## Desvantagem: Assintótico; menos confiável em n pequeno.
stat_dp, p_dp = st.normaltest(X)
print("D’Agostino–Pearson (Omnibus):")
print(f"  Estatística = {stat_dp:.4f}, p-valor = {p_dp:.4f}")
if p_dp < alpha:
    print("  Decisão: Rejeita H0 (não Normal)\n")
else:
    print("  Decisão: Não rejeita H0 (Normal)\n")
```

```python
# ---------------------------------------------------------------------------
#  Testes para Igualdade de Médias (paramétricos)
# ---------------------------------------------------------------------------
# Assumem (em geral) amostras ~ Normais. Para 2 grupos independentes:
# - Pooled: variâncias iguais
# - Welch: variâncias diferentes
# Para dados emparelhados: t pareado
# Para mais de 2 grupos: ANOVA (one-way)

import numpy as np
import pandas as pd
from scipy import stats as st

# ---------------------------------------------------------------------------
# Gerando bases de exemplo
# ---------------------------------------------------------------------------
rng   = np.random.default_rng(123)
alpha = 0.05

# Uma amostra (para Z e t de 1 amostra)
n1   = 50
mu_x = 10.0
sd_x = 2.0
X1   = rng.normal(mu_x, sd_x, size=n1)

# Parâmetros populacionais "conhecidos" (para o Z-test de média)
mu0    = 10.0          # média sob H0
sigma0 = 2.0           # σ populacional conhecido (não é o desvio amostral)

# Duas amostras independentes (para t independente)
nA, nB = 40, 45
muA, muB = 0.0, 0.5    # mude para criar diferenças
sdA, sdB = 1.0, 1.0    # iguais -> cenário pooled; mude sdB p/ 1.6 e use Welch
A = rng.normal(muA, sdA, size=nA)
B = rng.normal(muB, sdB, size=nB)

# Amostras pareadas (antes/depois)
nP     = 30
mu_pre = 5.0
sd_pre = 1.2
# Efeito "depois" (diferença média)
delta  = 0.6
PRE    = rng.normal(mu_pre, sd_pre, size=nP)
POST   = PRE + delta + rng.normal(0, 0.5, size=nP)

# Três grupos para ANOVA
nG = 35
G1 = rng.normal(0.0, 1.0, size=nG)
G2 = rng.normal(0.5, 1.0, size=nG)
G3 = rng.normal(1.0, 1.0, size=nG)

# ---------------------------------------------------------------------------
# Z-test para a média (σ conhecido)
# ---------------------------------------------------------------------------
## Usado quando: σ populacional é conhecido (raro na prática, comum em exercícios/teoria).
## Vantagem: Estatística Z com distribuição Normal padrão.
## Desvantagem: Na prática σ é desconhecido; usa-se t em vez de Z.
xbar = X1.mean()
Z    = (xbar - mu0) / (sigma0 / np.sqrt(n1))
# teste bicaudal
p_Z  = 2 * st.norm.sf(np.abs(Z))
print("Z-test (σ conhecido) - 1 amostra:")
print(f"  x̄ = {xbar:.4f}, Z = {Z:.4f}, p-valor = {p_Z:.4f}")
print("  Decisão:", "Rejeita H0 (μ ≠ μ0)\n" if p_Z < alpha else "Não rejeita H0 (μ = μ0)\n")

# ---------------------------------------------------------------------------
# t de Student - 1 amostra (σ desconhecido)
# ---------------------------------------------------------------------------
## Usado quando: quer testar média de uma população com σ desconhecido.
## Vantagem: Usa a t (df = n-1), padrão na prática.
## Desvantagem: Supõe Normalidade (ou n grande via TLC).
t_1samp, p_1samp = st.ttest_1samp(X1, popmean=mu0, alternative='two-sided')
print("t (1 amostra):")
print(f"  x̄ = {xbar:.4f}, t = {t_1samp:.4f}, df = {n1-1}, p-valor = {p_1samp:.4f}")
print("  Decisão:", "Rejeita H0 (μ ≠ μ0)\n" if p_1samp < alpha else "Não rejeita H0 (μ = μ0)\n")

# ---------------------------------------------------------------------------
# t para 2 médias independentes - variâncias iguais (pooled)
# ---------------------------------------------------------------------------
## Usado quando: duas amostras independentes e variâncias ≈ iguais.
## Vantagem: Mais potente se a suposição de variâncias iguais é válida.
## Desvantagem: Sensível à violação de homocedasticidade.
t_ind_pooled, p_ind_pooled = st.ttest_ind(A, B, equal_var=True, alternative='two-sided')
print("t (2 amostras independentes) - Pooled (variâncias iguais):")
print(f"  x̄A = {A.mean():.4f}, x̄B = {B.mean():.4f}, t = {t_ind_pooled:.4f}, p-valor = {p_ind_pooled:.4f}")
print("  Decisão:", "Rejeita H0 (μA = μB)\n" if p_ind_pooled < alpha else "Não rejeita H0 (μA = μB)\n")

# ---------------------------------------------------------------------------
# t de Welch - 2 médias independentes (variâncias diferentes)
# ---------------------------------------------------------------------------
## Usado quando: variâncias são diferentes ou suspeitas de serem diferentes.
## Vantagem: Mais robusto a heterocedasticidade (ajusta df).
## Desvantagem: Leve perda de poder se variâncias forem realmente iguais.
t_welch, p_welch = st.ttest_ind(A, B, equal_var=False, alternative='two-sided')
print("t de Welch (2 amostras independentes) - variâncias diferentes:")
print(f"  x̄A = {A.mean():.4f}, x̄B = {B.mean():.4f}, t = {t_welch:.4f}, p-valor = {p_welch:.4f}")
print("  Decisão:", "Rejeita H0 (μA = μB)\n" if p_welch < alpha else "Não rejeita H0 (μA = μB)\n")

# ---------------------------------------------------------------------------
# t pareado (duas amostras dependentes: antes/depois, medidas repetidas)
# ---------------------------------------------------------------------------
## Usado quando: pares naturais (mesmo indivíduo medido 2x).
## Vantagem: Controla variabilidade entre indivíduos (ganho de potência).
## Desvantagem: Requer emparelhamento verdadeiro; outliers nos deltas afetam.
t_paired, p_paired = st.ttest_rel(PRE, POST, alternative='two-sided')
print("t pareado (antes/depois):")
print(f"  x̄PRE = {PRE.mean():.4f}, x̄POST = {POST.mean():.4f}, t = {t_paired:.4f}, p-valor = {p_paired:.4f}")
print("  Decisão:", "Rejeita H0 (μPRE = μPOST)\n" if p_paired < alpha else "Não rejeita H0 (μPRE = μPOST)\n")

# ---------------------------------------------------------------------------
# ANOVA (one-way) - 3 grupos independentes
# ---------------------------------------------------------------------------
## Usado quando: comparar mais de 2 médias simultaneamente (controle do α global).
## Vantagem: Evita inflar erro tipo I de múltiplos t.
## Desvantagem: Supõe normalidade e homocedasticidade; se rejeitar, pós-testes (Tukey, etc.).
F_anova, p_anova = st.f_oneway(G1, G2, G3)
print("ANOVA (one-way):")
print(f"  F = {F_anova:.4f}, p-valor = {p_anova:.4f}")
print("  Decisão:", "Rejeita H0 (todas as médias iguais)\n" if p_anova < alpha else "Não rejeita H0 (médias iguais)\n")
```

```python
# ---------------------------------------------------------------------------
#  Testes para Variância (paramétricos)
# ---------------------------------------------------------------------------
# Assumem dados ~ Normais. São usados para verificar:
# - se a variância é igual a um valor específico (1 amostra)
# - se duas variâncias são iguais (2 amostras independentes)
# - se mais de duas variâncias são iguais (Bartlett)

import numpy as np
import pandas as pd
from scipy import stats as st

# ---------------------------------------------------------------------------
# Gerando bases de exemplo
# ---------------------------------------------------------------------------
rng   = np.random.default_rng(2025)
alpha = 0.05

# Uma amostra para teste de variância (média conhecida/desconhecida)
n1   = 25
mu_x = 50.0
sigma_x = 5.0
X1   = rng.normal(mu_x, sigma_x, size=n1)

# Valor de σ² sob H0
sigma2_0 = 25.0  # (5²)

# Duas amostras independentes para comparação de variâncias
nA, nB = 20, 22
A = rng.normal(10.0, 3.0, size=nA)
B = rng.normal(12.0, 4.0, size=nB)

# Três grupos para Bartlett
nG = 18
G1 = rng.normal(0.0, 2.0, size=nG)
G2 = rng.normal(1.0, 2.5, size=nG)
G3 = rng.normal(0.5, 2.0, size=nG)

# ---------------------------------------------------------------------------
# Qui-quadrado para variância — média conhecida
# ---------------------------------------------------------------------------
## Usado quando: média populacional é conhecida e se quer testar σ² = σ₀².
## Vantagem: Simples e direto quando μ é fixo.
## Desvantagem: Raro ter μ populacional conhecido na prática.
mu0 = 50.0  # média conhecida
SS  = np.sum((X1 - mu0)**2)  # soma de quadrados em torno de μ conhecido
chi2_stat = SS / sigma2_0
df = n1  # graus de liberdade = n (pois μ é conhecido)
p_chi2 = 2 * min(st.chi2.cdf(chi2_stat, df), 1 - st.chi2.cdf(chi2_stat, df))
print("Qui-quadrado (média conhecida):")
print(f"  Estatística = {chi2_stat:.4f}, gl = {df}, p-valor = {p_chi2:.4f}")
print("  Decisão:", "Rejeita H0 (σ² ≠ σ₀²)\n" if p_chi2 < alpha else "Não rejeita H0 (σ² = σ₀²)\n")

# ---------------------------------------------------------------------------
# Qui-quadrado para variância — média desconhecida
# ---------------------------------------------------------------------------
## Usado quando: média populacional é desconhecida e se quer testar σ² = σ₀².
## Vantagem: Mais comum na prática que o caso anterior.
## Desvantagem: Supõe normalidade; gl = n-1.
s2 = np.var(X1, ddof=1)
chi2_stat2 = (n1 - 1) * s2 / sigma2_0
df2 = n1 - 1
p_chi2_2 = 2 * min(st.chi2.cdf(chi2_stat2, df2), 1 - st.chi2.cdf(chi2_stat2, df2))
print("Qui-quadrado (média desconhecida):")
print(f"  Estatística = {chi2_stat2:.4f}, gl = {df2}, p-valor = {p_chi2_2:.4f}")
print("  Decisão:", "Rejeita H0 (σ² ≠ σ₀²)\n" if p_chi2_2 < alpha else "Não rejeita H0 (σ² = σ₀²)\n")

# ---------------------------------------------------------------------------
# F-test para igualdade de variâncias — 2 amostras independentes
# ---------------------------------------------------------------------------
## Usado quando: quer comparar σ² de duas populações independentes.
## Vantagem: Simples; base para ANOVA.
## Desvantagem: Muito sensível a desvios da normalidade.
varA = np.var(A, ddof=1)
varB = np.var(B, ddof=1)
F_stat = varA / varB
df1, df2 = len(A) - 1, len(B) - 1
p_F = 2 * min(st.f.cdf(F_stat, df1, df2), 1 - st.f.cdf(F_stat, df1, df2))
print("F-test (igualdade de duas variâncias):")
print(f"  Estatística = {F_stat:.4f}, gl = ({df1},{df2}), p-valor = {p_F:.4f}")
print("  Decisão:", "Rejeita H0 (σA² = σB²)\n" if p_F < alpha else "Não rejeita H0 (σA² = σB²)\n")

# ---------------------------------------------------------------------------
# Bartlett — igualdade de variâncias (mais de 2 grupos)
# ---------------------------------------------------------------------------
## Usado quando: quer comparar variâncias de k grupos independentes.
## Vantagem: Generalização do F-test para k>2.
## Desvantagem: Muito sensível à não-normalidade (use Levene/Brown-Forsythe se duvidar da normalidade).
stat_bartlett, p_bartlett = st.bartlett(G1, G2, G3)
print("Bartlett (igualdade de k variâncias):")
print(f"  Estatística = {stat_bartlett:.4f}, p-valor = {p_bartlett:.4f}")
print("  Decisão:", "Rejeita H0 (todas as variâncias iguais)\n" if p_bartlett < alpha else "Não rejeita H0 (variâncias iguais)\n")
```

```python
# ---------------------------------------------------------------------------
#  Testes para Proporções (paramétricos)
# ---------------------------------------------------------------------------
# Assumem amostras binárias independentes e n grande o suficiente
# para usar aproximação Normal à Binomial.
# Usados para comparar proporções com um valor hipotético ou entre grupos.

import numpy as np
from scipy import stats as st

# ---------------------------------------------------------------------------
# Gerando bases de exemplo
# ---------------------------------------------------------------------------
alpha = 0.05

# 1 amostra binária — exemplo: 40 sucessos em 100 tentativas
x_sucessos = 40
n_total    = 100
p0         = 0.5  # proporção sob H0

# Duas amostras — exemplo: campanha A e campanha B
xA, nA = 30, 80
xB, nB = 25, 90

# ---------------------------------------------------------------------------
# Teste Z para 1 proporção
# ---------------------------------------------------------------------------
## Usado quando: quer comparar uma proporção amostral p̂ com um valor p0 conhecido.
## Vantagem: Simples; interpretação direta com Normal padrão.
## Desvantagem: Aproximação ruim se n pequeno ou p próximo de 0/1.
p_hat = x_sucessos / n_total
Z_1prop = (p_hat - p0) / np.sqrt(p0 * (1 - p0) / n_total)
p_1prop = 2 * st.norm.sf(np.abs(Z_1prop))  # bicaudal
print("Teste Z - 1 proporção:")
print(f"  p̂ = {p_hat:.4f}, Z = {Z_1prop:.4f}, p-valor = {p_1prop:.4f}")
print("  Decisão:", "Rejeita H0 (p ≠ p0)\n" if p_1prop < alpha else "Não rejeita H0 (p = p0)\n")

# ---------------------------------------------------------------------------
# Teste Z para 2 proporções independentes
# ---------------------------------------------------------------------------
## Usado quando: quer comparar duas proporções p̂1 e p̂2 de amostras independentes.
## Vantagem: Simples; muito usado em A/B testing.
## Desvantagem: Aproximação ruim se n pequeno ou p próximo de 0/1.
p1_hat = xA / nA
p2_hat = xB / nB
p_pooled = (xA + xB) / (nA + nB)
Z_2prop = (p1_hat - p2_hat) / np.sqrt(p_pooled * (1 - p_pooled) * (1/nA + 1/nB))
p_2prop = 2 * st.norm.sf(np.abs(Z_2prop))
print("Teste Z - 2 proporções independentes:")
print(f"  p̂1 = {p1_hat:.4f}, p̂2 = {p2_hat:.4f}, Z = {Z_2prop:.4f}, p-valor = {p_2prop:.4f}")
print("  Decisão:", "Rejeita H0 (p1 ≠ p2)\n" if p_2prop < alpha else "Não rejeita H0 (p1 = p2)\n")

# ---------------------------------------------------------------------------
# Teste qui-quadrado para k proporções (tabela de contingência)
# ---------------------------------------------------------------------------
## Usado quando: quer comparar proporções entre mais de 2 grupos ou categorias.
## Vantagem: Generaliza o teste de 2 proporções; aplica-se a tabelas rxc.
## Desvantagem: Exige frequências esperadas ≥ 5 (ou usar teste exato de Fisher).
# Exemplo: sucesso/fracasso para 3 grupos
sucessos = np.array([30, 25, 20])
fracassos = np.array([50, 55, 60])
tabela = np.vstack([sucessos, fracassos]).T  # linhas = grupos, colunas = categorias

chi2_stat, p_chi2, dof, exp = st.chi2_contingency(tabela)
print("Qui-quadrado para k proporções:")
print(f"  Estatística = {chi2_stat:.4f}, gl = {dof}, p-valor = {p_chi2:.4f}")
print("  Decisão:", "Rejeita H0 (todas as proporções iguais)\n" if p_chi2 < alpha else "Não rejeita H0 (proporções iguais)\n")
```

```python
# ---------------------------------------------------------------------------
#  Testes de Homocedasticidade (paramétricos e robustos)
# ---------------------------------------------------------------------------
# Usados para verificar se as variâncias de dois ou mais grupos são iguais.
# Relevantes para ANOVA e t-test (pooled).
# Paramétricos assumem normalidade (ex.: Bartlett, F-test).
# Levene e Brown–Forsythe são mais robustos a desvios da normalidade.

import numpy as np
from scipy import stats as st

# ---------------------------------------------------------------------------
# Gerando bases de exemplo
# ---------------------------------------------------------------------------
rng   = np.random.default_rng(2025)
alpha = 0.05

# Dois grupos para F-test
nA, nB = 20, 22
A = rng.normal(10.0, 3.0, size=nA)
B = rng.normal(12.0, 4.0, size=nB)

# Três grupos para Bartlett, Levene, Brown–Forsythe
nG = 18
G1 = rng.normal(0.0, 2.0, size=nG)
G2 = rng.normal(1.0, 2.5, size=nG)
G3 = rng.normal(0.5, 2.0, size=nG)

# ---------------------------------------------------------------------------
# F-test para igualdade de duas variâncias
# ---------------------------------------------------------------------------
## Usado quando: quer comparar σ² de duas populações independentes.
## Vantagem: Simples e direto.
## Desvantagem: Muito sensível a desvios da normalidade.
varA = np.var(A, ddof=1)
varB = np.var(B, ddof=1)
F_stat = varA / varB
df1, df2 = len(A) - 1, len(B) - 1
p_F = 2 * min(st.f.cdf(F_stat, df1, df2), 1 - st.f.cdf(F_stat, df1, df2))
print("F-test (igualdade de duas variâncias):")
print(f"  Estatística = {F_stat:.4f}, gl = ({df1},{df2}), p-valor = {p_F:.4f}")
print("  Decisão:", "Rejeita H0 (σA² = σB²)\n" if p_F < alpha else "Não rejeita H0 (σA² = σB²)\n")

# ---------------------------------------------------------------------------
# Bartlett — igualdade de variâncias (k grupos)
# ---------------------------------------------------------------------------
## Usado quando: quer comparar variâncias de mais de dois grupos.
## Vantagem: Muito usado em ANOVA.
## Desvantagem: Muito sensível à não-normalidade.
stat_bartlett, p_bartlett = st.bartlett(G1, G2, G3)
print("Bartlett (igualdade de k variâncias):")
print(f"  Estatística = {stat_bartlett:.4f}, p-valor = {p_bartlett:.4f}")
print("  Decisão:", "Rejeita H0 (todas as variâncias iguais)\n" if p_bartlett < alpha else "Não rejeita H0 (variâncias iguais)\n")

# ---------------------------------------------------------------------------
# Levene — robusto a desvios de normalidade
# ---------------------------------------------------------------------------
## Usado quando: quer comparar variâncias sem assumir normalidade estrita.
## Vantagem: Mais robusto que Bartlett; usa desvios em relação à média.
## Desvantagem: Menos potente se normalidade for atendida perfeitamente.
stat_levene, p_levene = st.levene(G1, G2, G3, center='mean')
print("Levene (center = mean):")
print(f"  Estatística = {stat_levene:.4f}, p-valor = {p_levene:.4f}")
print("  Decisão:", "Rejeita H0 (todas as variâncias iguais)\n" if p_levene < alpha else "Não rejeita H0 (variâncias iguais)\n")

# ---------------------------------------------------------------------------
# Brown–Forsythe — robusto (variante do Levene usando a mediana)
# ---------------------------------------------------------------------------
## Usado quando: quer ainda mais robustez a outliers e assimetria.
## Vantagem: Usa desvios em relação à mediana; ótimo quando há assimetria.
## Desvantagem: Pouco conhecido fora de estatística aplicada.
stat_bf, p_bf = st.levene(G1, G2, G3, center='median')
print("Brown–Forsythe (center = median):")
print(f"  Estatística = {stat_bf:.4f}, p-valor = {p_bf:.4f}")
print("  Decisão:", "Rejeita H0 (todas as variâncias iguais)\n" if p_bf < alpha else "Não rejeita H0 (variâncias iguais)\n")
```

```python
# ---------------------------------------------------------------------------
#  Testes Qui-Quadrado
# ---------------------------------------------------------------------------
# Usados para analisar dados categóricos, comparando frequências observadas
# com frequências esperadas (aderência) ou verificando associações/igualdade
# de distribuições entre grupos (homogeneidade, independência).

import numpy as np
from scipy import stats as st

alpha = 0.05

# ---------------------------------------------------------------------------
# 1. Teste de Aderência (Goodness-of-Fit)
# ---------------------------------------------------------------------------
## Usado quando: quer verificar se a distribuição observada segue um padrão teórico.
## Vantagem: Simples; compara com frequências esperadas.
## Desvantagem: Frequências esperadas devem ser ≥ 5 (ou usar versões exatas).

# Exemplo: lançamento de um dado — queremos verificar se é justo
obs = np.array([10, 15, 14, 16, 18, 7])          # frequências observadas
n_total = obs.sum()
esp = np.array([n_total/6]*6)                    # frequências esperadas para dado justo

chi2_stat, p_chi2 = st.chisquare(obs, f_exp=esp)
print("Qui-quadrado de Aderência:")
print(f"  Estatística = {chi2_stat:.4f}, gl = {len(obs)-1}, p-valor = {p_chi2:.4f}")
print("  Decisão:", "Rejeita H0 (não segue a distribuição esperada)\n" if p_chi2 < alpha else "Não rejeita H0 (segue a distribuição esperada)\n")

# ---------------------------------------------------------------------------
# 2. Teste de Homogeneidade
# ---------------------------------------------------------------------------
## Usado quando: quer verificar se duas ou mais populações têm a mesma distribuição categórica.
## Vantagem: Permite comparar padrões de frequência entre grupos.
## Desvantagem: Requer amostras independentes; frequências esperadas ≥ 5.

# Exemplo: preferência por sabores em 3 cidades
tabela_homog = np.array([
    [30, 20, 25],  # Cidade 1
    [25, 22, 28],  # Cidade 2
    [28, 18, 24]   # Cidade 3
])
chi2_stat_h, p_chi2_h, dof_h, exp_h = st.chi2_contingency(tabela_homog)
print("Qui-quadrado de Homogeneidade:")
print(f"  Estatística = {chi2_stat_h:.4f}, gl = {dof_h}, p-valor = {p_chi2_h:.4f}")
print("  Decisão:", "Rejeita H0 (distribuições diferentes entre grupos)\n" if p_chi2_h < alpha else "Não rejeita H0 (distribuições iguais entre grupos)\n")

# ---------------------------------------------------------------------------
# 3. Teste de Independência (tabela de contingência)
# ---------------------------------------------------------------------------
## Usado quando: quer verificar se duas variáveis categóricas são independentes.
## Vantagem: Muito usado em análise de tabelas cruzadas.
## Desvantagem: Requer frequências esperadas ≥ 5 (ou usar teste exato de Fisher para tabelas pequenas).

# Exemplo: gênero vs. aprovação em teste
tabela_indep = np.array([
    [30, 20],  # Masculino (Aprovado, Reprovado)
    [25, 25]   # Feminino  (Aprovado, Reprovado)
])
chi2_stat_i, p_chi2_i, dof_i, exp_i = st.chi2_contingency(tabela_indep)
print("Qui-quadrado de Independência:")
print(f"  Estatística = {chi2_stat_i:.4f}, gl = {dof_i}, p-valor = {p_chi2_i:.4f}")
print("  Decisão:", "Rejeita H0 (variáveis associadas)\n" if p_chi2_i < alpha else "Não rejeita H0 (variáveis independentes)\n")
```
